```python
from __future__ import annotations
import scanpy as sc, anndata as ad, pandas as pd, numpy as np
from pathlib import Path

# ---- your absolute paths ----
RAW_10X  = Path("/Users/david-johnlonge/Documents/Uni stuff/diss/8273102/tabula-muris-senis-droplet-official-raw-obj.h5ad")
RAW_SS2  = Path("/Users/david-johnlonge/Documents/Uni stuff/diss/8273102/tabula-muris-senis-facs-official-raw-obj.h5ad")
PROC_10X = Path("/Users/david-johnlonge/Documents/Uni stuff/diss/8273102/tabula-muris-senis-droplet-processed-official-annotations.h5ad")
PROC_SS2 = Path("/Users/david-johnlonge/Documents/Uni stuff/diss/8273102/tabula-muris-senis-facs-processed-official-annotations.h5ad")

OUTDIR   = Path("/Users/david-johnlonge/Documents/Uni stuff/diss/8273102/standardised")
TABDIR   = Path("/Users/david-johnlonge/Documents/Uni stuff/diss/8273102/results/tables")
OUTDIR.mkdir(parents=True, exist_ok=True)
TABDIR.mkdir(parents=True, exist_ok=True)

def _parse_age_to_months(x):
    if pd.isna(x): return np.nan
    s = str(x).strip().lower()
    if s.endswith("m"):
        try: return float(s[:-1])
        except: return np.nan
    try: return float(s)
    except: return np.nan

def _age_group(m):
    if pd.isna(m): return None
    if m <= 3.5: return "young"
    if m >= 18.0: return "old"
    return None  # keep strictly young vs old

def _add_annotations_from_processed(ad_raw: ad.AnnData, proc_path: Path, platform: str) -> ad.AnnData:
    ad_proc = sc.read_h5ad(proc_path, backed="r")
    obs_p = ad_proc.obs.copy()
    ad_proc.file.close()

    common = ad_raw.obs_names.intersection(obs_p.index)
    if common.size == 0:
        raise ValueError(f"No overlapping cell IDs between RAW and PROCESSED for {platform}")

    ad_raw = ad_raw[common].copy()
    obs_p  = obs_p.loc[common]

    tissue_key = next((k for k in ["tissue","organ","tissue_ontology_label"] if k in obs_p), None)
    cell_key   = next((k for k in ["cell_ontology_class","cell_ontology_label","cell_type","free_annotation"] if k in obs_p), None)
    age_key    = next((k for k in ["age","age_group","age_label"] if k in obs_p), None)
    batch_key  = next((k for k in ["batch","donor","mouse.id","sample","indiv"] if k in obs_p), None)
    if not all([tissue_key, cell_key, age_key]):
        raise KeyError(f"Missing required keys in processed {platform}. Got: {list(obs_p.columns)[:25]}")

    ad_raw.obs["tissue"]     = obs_p[tissue_key].astype(str)
    ad_raw.obs["cell_type"]  = obs_p[cell_key].astype(str)
    ad_raw.obs["age_months"] = obs_p[age_key].map(_parse_age_to_months)
    ad_raw.obs["age_group"]  = ad_raw.obs["age_months"].map(_age_group)
    ad_raw.obs["batch"]      = obs_p[batch_key].astype(str) if batch_key else "NA"
    ad_raw.obs["platform"]   = platform

    if "counts" not in ad_raw.layers:
        ad_raw.layers["counts"] = ad_raw.X.copy()

    # keep a human-friendly gene symbol
    if "gene_symbols" in ad_raw.var:
        ad_raw.var["gene_symbol"] = ad_raw.var["gene_symbols"].astype(str)
    else:
        ad_raw.var["gene_symbol"] = ad_raw.var_names.astype(str)
    ad_raw.var_names_make_unique()

    # keep only young/old
    ad_raw = ad_raw[ad_raw.obs["age_group"].notna()].copy()
    return ad_raw

print("Loading raw 10x…")
ad10x = sc.read_h5ad(RAW_10X)
ad10x = _add_annotations_from_processed(ad10x, PROC_10X, "10x")
print(ad10x)

print("Loading raw Smart-seq2…")
adss2 = sc.read_h5ad(RAW_SS2)
adss2 = _add_annotations_from_processed(adss2, PROC_SS2, "Smart-seq2")
print(adss2)

# Intersect genes (prefer exact var_names; fallback via gene_symbol)
common = ad10x.var_names.intersection(adss2.var_names)
if common.size == 0:
    a = ad10x.var.reset_index().rename(columns={"index":"vn_10x"})
    b = adss2.var.reset_index().rename(columns={"index":"vn_ss2"})
    m = a.merge(b, on="gene_symbol", how="inner")
    ad10x = ad10x[:, m["vn_10x"].values].copy()
    adss2 = adss2[:, m["vn_ss2"].values].copy()
    ad10x.var_names = ad10x.var["gene_symbol"].astype(str)
    adss2.var_names = adss2.var["gene_symbol"].astype(str)
    ad10x.var_names_make_unique(); adss2.var_names_make_unique()
else:
    ad10x = ad10x[:, common].copy()
    adss2 = adss2[:, common].copy()

def strata_summary(adata: ad.AnnData) -> pd.DataFrame:
    return (adata.obs.groupby(["tissue","cell_type","age_group","platform"])
            .size().rename("n_cells").reset_index()
            .sort_values("n_cells", ascending=False))

sum10x = strata_summary(ad10x)
sumss2 = strata_summary(adss2)

print("\nTop 12 strata 10x\n", sum10x.head(12).to_string(index=False))
print("\nTop 12 strata SS2\n", sumss2.head(12).to_string(index=False))

# Save standardised copies for the rest of the pipeline
ad10x.write_h5ad(OUTDIR / "tms_10x_std.h5ad", compression="lzf")
adss2.write_h5ad(OUTDIR / "tms_ss2_std.h5ad", compression="lzf")
sum10x.to_csv(TABDIR / "strata_10x.csv", index=False)
sumss2.to_csv(TABDIR / "strata_ss2.csv", index=False)
print("\nSaved:\n ", OUTDIR / "tms_10x_std.h5ad", "\n ", OUTDIR / "tms_ss2_std.h5ad")
```

```python
import re, pandas as pd, scanpy as sc
from pathlib import Path

STD_10X = Path("/Users/david-johnlonge/Documents/Uni stuff/diss/8273102/standardised/tms_10x_std.h5ad")
STD_SS2 = Path("/Users/david-johnlonge/Documents/Uni stuff/diss/8273102/standardised/tms_ss2_std.h5ad")
OUTCSV  = Path("/Users/david-johnlonge/Documents/Uni stuff/diss/8273102/results/tables/matched_strata.csv")

ad10x = sc.read_h5ad(STD_10X)
adss2 = sc.read_h5ad(STD_SS2)

# --- tissue harmonisation (expand as needed) ---
TISSUE_MAP = {
    "Brain_Myeloid": "Brain",
    "Brain_Non-Myeloid": "Brain",
    "Heart": "Heart_and_Aorta",
    "Aorta": "Heart_and_Aorta",
    "BAT": "Fat", "GAT": "Fat", "MAT": "Fat", "SCAT": "Fat",
    "Diaphragm": "Limb_Muscle",
}
for ad in (ad10x, adss2):
    ad.obs["tissue_std"] = ad.obs["tissue"].map(lambda s: TISSUE_MAP.get(str(s), str(s)))

# --- super-type mapper (coarsen cell ontology class/label) ---
SUPER_PATTERNS = [
    (r"(naive\s+)?b\s+cell|b\s+cell",                            "B cell"),
    (r"(cd4|cd8).*\bt\s+cell|t\s+cell|t-cell",                   "T cell"),
    (r"granulocyte|neutrophil|granulocytopoietic",               "Granulocyte"),
    (r"monocyte|macrophage|dendritic|kupffer|microglia",         "Mononuclear phagocyte"),
    (r"endothelial",                                             "Endothelial cell"),
    (r"fibroblast|stromal",                                      "Fibroblast"),
    (r"keratinocyte|basal cell of epidermis|epidermal",          "Keratinocyte"),
    (r"urothelial",                                              "Urothelial cell"),
    (r"hepatocyte",                                              "Hepatocyte"),
    (r"alveolar.*type\s*2|type\s*ii.*pneumocyte|at2",            "AT2 cell"),
    (r"pericyte",                                                "Pericyte"),
    (r"smooth\s*muscle",                                         "Smooth muscle"),
    (r"hematopoietic stem cell|hsc\b",                           "HSC"),
    (r"erythro",                                                 "Erythroid"),
]

def to_supertype(label: str) -> str:
    s = str(label).lower()
    for pat, sup in SUPER_PATTERNS:
        if re.search(pat, s):
            return sup
    return label  # fallback: keep original

def add_supertype_col(ad):
    lab = "cell_ontology_class" if "cell_ontology_class" in ad.obs else "cell_type"
    ad.obs["cell_supertype"] = ad.obs[lab].map(to_supertype)

add_supertype_col(ad10x); add_supertype_col(adss2)

# keep only tissues present in both
both_tissues = set(ad10x.obs["tissue_std"]).intersection(set(adss2.obs["tissue_std"]))
ad10x = ad10x[ad10x.obs["tissue_std"].isin(both_tissues)].copy()
adss2 = adss2[adss2.obs["tissue_std"].isin(both_tissues)].copy()

# counts per tissue × supertype × age × platform
def strata_counts(ad):
    return (ad.obs.groupby(["tissue_std","cell_supertype","age_group","platform"], observed=True)
              .size().rename("n").reset_index())

c10 = strata_counts(ad10x)
css = strata_counts(adss2)
df = pd.concat([c10, css], ignore_index=True)

# pivot and threshold
tab = (df.pivot_table(index=["tissue_std","cell_supertype"],
                      columns=["platform","age_group"],
                      values="n", fill_value=0, observed=True)
         .reset_index())

MIN_10X = 50
MIN_SS2 = 25  # SS2 is sparser; 25 per age is a pragmatic floor

def ok(r, plat, age, thr):
    col = (plat, age)
    return (col in r.index) and (r[col] >= thr)

keep = tab.apply(lambda r:
                 ok(r, "10x", "young", MIN_10X) and ok(r, "10x", "old", MIN_10X) and
                 ok(r, "Smart-seq2", "young", MIN_SS2) and ok(r, "Smart-seq2", "old", MIN_SS2),
                 axis=1)

matched = tab[keep].copy()
if matched.empty:
    # last resort: allow SS2 ≥ 20
    keep = tab.apply(lambda r:
                     ok(r, "10x", "young", MIN_10X) and ok(r, "10x", "old", MIN_10X) and
                     ok(r, "Smart-seq2", "young", 20) and ok(r, "Smart-seq2", "old", 20),
                     axis=1)
    matched = tab[keep].copy()

# sort and save
if not matched.empty:
    matched["total_cells"] = 0
    for col in matched.columns:
        if isinstance(col, tuple) and col and col[0] in ("10x","Smart-seq2"):
            matched["total_cells"] += matched[col]
    matched = matched.sort_values("total_cells", ascending=False)
matched.to_csv(OUTCSV, index=False)
print("Matched strata (supertype) saved:", OUTCSV)
print(matched.head(30).to_string(index=False))
```

```python
import numpy as np
from scipy import sparse

def add_mito_fraction(ad):
    # mouse mitochondrial genes often start with 'mt-'
    if "gene_symbol" in ad.var:
        sym = ad.var["gene_symbol"].astype(str).str.lower()
    else:
        sym = pd.Index(ad.var_names.astype(str)).str.lower()
    mito = sym.str.startswith(("mt-","mt_"))
    if mito.any():
        X = ad.layers.get("counts", ad.X)
        if sparse.issparse(X):
            total = np.asarray(X.sum(axis=1)).ravel()
            mito_counts = np.asarray(X[:, mito].sum(axis=1)).ravel()
        else:
            total = X.sum(axis=1)
            mito_counts = X[:, mito].sum(axis=1)
        ad.obs["pct_counts_mt"] = np.where(total>0, mito_counts / total * 100.0, 0.0)
    else:
        ad.obs["pct_counts_mt"] = np.nan

def qc_filter(ad, platform: str):
    add_mito_fraction(ad)
    # per-cell thresholds (light-touch; adjust if needed)
    if platform == "10x":
        keep = (ad.obs["n_genes"] >= 200) & (ad.obs["pct_counts_mt"].fillna(0) <= 20)
    else:  # Smart-seq2 (full-length)
        # SS2 has more genes/cell; also expect higher mito fraction variability
        keep = (ad.obs["n_genes"] >= 500) & (ad.obs["pct_counts_mt"].fillna(0) <= 30)
    return ad[keep].copy()

def gene_filter(counts, min_detect_frac=0.10, min_mean=0.01):
    # counts: 2D array (cells x genes), return boolean mask for genes to keep
    n_cells = counts.shape[0]
    detect = (counts > 0).sum(axis=0) / max(1, n_cells)
    mean = counts.mean(axis=0)
    return (detect >= min_detect_frac) & (mean >= min_mean)
```

```python
import numpy as np
import pandas as pd

def moments_nb(counts):
    """
    counts: 2D array (cells x genes) of raw counts for a stratum.
    Returns mu, var, a_hat, b_hat; masks invalid genes where var<=mu.
    """
    if sparse.issparse(counts):
        counts = counts.toarray()
    mu = counts.mean(axis=0)
    var = counts.var(axis=0, ddof=0)
    over = var > mu  # NB-like overdispersion
    a_hat = np.where(over, (mu**2) / (var - mu + 1e-12), np.nan)
    b_hat = np.where(over, (var - mu) / (mu + 1e-12),     np.nan)
    return mu, var, a_hat, b_hat, over

def log2fc(x_old, x_yng, eps=1e-6):
    return np.log2((x_old + eps) / (x_yng + eps))

def bootstrap_delta(counts_old, counts_yng, n_boot=500, seed=42):
    """
    Bootstrap CIs for delta(log2 a_hat) and delta(log2 b_hat).
    Returns dict with lo/hi percentiles and a two-sided p-value by sign proportion.
    """
    rng = np.random.default_rng(seed)
    n_o, n_y = counts_old.shape[0], counts_yng.shape[0]
    if sparse.issparse(counts_old): counts_old = counts_old.toarray()
    if sparse.issparse(counts_yng): counts_yng = counts_yng.toarray()
    deltas_a, deltas_b = [], []
    for _ in range(n_boot):
        idx_o = rng.integers(0, n_o, size=n_o)
        idx_y = rng.integers(0, n_y, size=n_y)
        mu_o, var_o, a_o, b_o, over_o = moments_nb(counts_old[idx_o, :])
        mu_y, var_y, a_y, b_y, over_y = moments_nb(counts_yng[idx_y, :])
        ok = over_o & over_y & np.isfinite(a_o) & np.isfinite(a_y) & (a_o>0) & (a_y>0) & (b_o>=0) & (b_y>=0)
        if ok.any():
            deltas_a.append(np.nan_to_num(np.log2(a_o[ok]+1e-6) - np.log2(a_y[ok]+1e-6)))
            deltas_b.append(np.nan_to_num(np.log2(b_o[ok]+1e-6) - np.log2(b_y[ok]+1e-6)))
        else:
            deltas_a.append(np.array([]))
            deltas_b.append(np.array([]))
    # Aggregate by median over genes for a simple stratum-level p (optional)
    # We also compute per-gene CI below; this stratum-level piece is mostly for QA.
    return deltas_a, deltas_b

def per_gene_bootstrap(counts_old, counts_yng, n_boot=300, seed=123):
    """
    Per-gene bootstrap CIs and p-values for delta(log2 a_hat) and delta(log2 b_hat).
    """
    rng = np.random.default_rng(seed)
    n_o, n_y = counts_old.shape[0], counts_yng.shape[0]
    if sparse.issparse(counts_old): counts_old = counts_old.toarray()
    if sparse.issparse(counts_yng): counts_yng = counts_yng.toarray()
    G = counts_old.shape[1]
    # baseline estimates
    mu_o, var_o, a_o, b_o, ok_o = moments_nb(counts_old)
    mu_y, var_y, a_y, b_y, ok_y = moments_nb(counts_yng)
    ok = ok_o & ok_y & np.isfinite(a_o) & np.isfinite(a_y) & (a_o>0) & (a_y>0) & (b_o>=0) & (b_y>=0)
    l2fc_a = np.full(G, np.nan); l2fc_b = np.full(G, np.nan)
    l2fc_a[ok] = log2fc(a_o[ok], a_y[ok]); l2fc_b[ok] = log2fc(b_o[ok], b_y[ok])

    # bootstrap distributions per gene (vectorised loops)
    B = n_boot
    boot_a = np.full((B, G), np.nan, dtype=float)
    boot_b = np.full((B, G), np.nan, dtype=float)
    for b in range(B):
        io = rng.integers(0, n_o, size=n_o)
        iy = rng.integers(0, n_y, size=n_y)
        _, _, a_ob, b_ob, ok_ob = moments_nb(counts_old[io, :])
        _, _, a_yb, b_yb, ok_yb = moments_nb(counts_yng[iy, :])
        okb = ok_ob & ok_yb & np.isfinite(a_ob) & np.isfinite(a_yb) & (a_ob>0) & (a_yb>0) & (b_ob>=0) & (b_yb>=0)
        # only store for genes usable in this bootstrap draw
        idx = np.where(okb)[0]
        if idx.size:
            boot_a[b, idx] = np.log2(a_ob[idx]+1e-6) - np.log2(a_yb[idx]+1e-6)
            boot_b[b, idx] = np.log2(b_ob[idx]+1e-6) - np.log2(b_yb[idx]+1e-6)

    # CIs and p-values (two-sided, sign-based)
    def cis_p(boot):
        lo = np.nanpercentile(boot, 2.5, axis=0)
        hi = np.nanpercentile(boot, 97.5, axis=0)
        med = np.nanmedian(boot, axis=0)
        # p ≈ 2 * min(Pr(delta<=0), Pr(delta>=0))
        p_lo = np.nanmean(boot <= 0, axis=0)
        p_hi = np.nanmean(boot >= 0, axis=0)
        p = 2*np.minimum(p_lo, p_hi)
        return med, lo, hi, p

    med_a, lo_a, hi_a, p_a = cis_p(boot_a)
    med_b, lo_b, hi_b, p_b = cis_p(boot_b)

    res = pd.DataFrame({
        "gene": ad10x.var_names,  # both layers were intersected earlier
        "mu_old": mu_o, "var_old": var_o, "a_old": a_o, "b_old": b_o,
        "mu_yng": mu_y, "var_yng": var_y, "a_yng": a_y, "b_yng": b_y,
        "log2FC_a": l2fc_a, "log2FC_b": l2fc_b,
        "boot_med_a": med_a, "boot_lo_a": lo_a, "boot_hi_a": hi_a, "p_a": p_a,
        "boot_med_b": med_b, "boot_lo_b": lo_b, "boot_hi_b": hi_b, "p_b": p_b,
    })
    # BH-FDR per parameter
    for pcol, qcol in [("p_a","q_a"), ("p_b","q_b")]:
        pvals = res[pcol].to_numpy(dtype=float)
        order = np.argsort(np.nan_to_num(pvals, nan=1.0))
        q = np.full_like(pvals, np.nan, dtype=float)
        m = np.isfinite(pvals).sum()
        ranks = np.empty_like(order); ranks[order] = np.arange(1, len(order)+1)
        q[order] = pvals[order] * m / ranks[order]
        # monotone
        for i in range(len(q)-2, -1, -1):
            if np.isfinite(q[i]) and np.isfinite(q[i+1]):
                q[i] = min(q[i], q[i+1])
        res[qcol] = q
    return res
```

```python
import numpy as np
import scanpy as sc
from scipy import sparse

def per_cell_depth(X):
    if sparse.issparse(X): 
        return np.asarray(X.sum(axis=1)).ravel()
    return X.sum(axis=1)

def downsample_counts_matrix(X, target):
    """Multinomial thinning to exactly target counts per cell if possible; returns same type as input (CSR dense)."""
    import numpy as np
    from scipy import sparse
    if sparse.issparse(X):
        X = X.tocsr()
        rows = []
        for i in range(X.shape[0]):
            row = X.getrow(i).toarray().ravel()
            tot = row.sum()
            if tot <= target or tot == 0:
                rows.append(row)  # already low; keep as-is
            else:
                p = row / tot
                rows.append(np.random.multinomial(target, p))
        Xd = sparse.csr_matrix(np.vstack(rows))
        return Xd
    else:
        Xd = np.empty_like(X)
        for i in range(X.shape[0]):
            row = X[i]
            tot = row.sum()
            if tot <= target or tot == 0:
                Xd[i] = row
            else:
                p = row / tot
                Xd[i] = np.random.multinomial(target, p)
        return Xd

def matched_depth_counts(X_y, X_o, platform):
    """Return (Xy_ds, Xo_ds) downsampled to a common target depth within this platform/stratum."""
    dy = per_cell_depth(X_y); do = per_cell_depth(X_o)
    # robust target: min of the medians (avoid being dragged by a few low-depth cells)
    target = int(max(1, min(np.median(dy), np.median(do))))
    # for SS2, counts can be higher/fragmentary; still do the same thinning for fairness
    Xy_ds = downsample_counts_matrix(X_y, target)
    Xo_ds = downsample_counts_matrix(X_o, target)
    return Xy_ds, Xo_ds, target
```

```python
import numpy as np
from scipy import sparse

def per_cell_depth(X):
    if sparse.issparse(X):
        return np.asarray(X.sum(axis=1)).ravel()
    return X.sum(axis=1)

def downsample_counts_matrix_binomial(X, target):
    """
    Binomial thinning per gene to approximately match a per-cell target depth.
    Works with CSR sparse and dense arrays. Casts to integers safely.
    """
    if sparse.issparse(X):
        X = X.tocsr(copy=True)
        data = X.data
        indptr = X.indptr
        for i in range(X.shape[0]):
            s, e = indptr[i], indptr[i+1]
            if e <= s:  # empty row
                continue
            row_sum = data[s:e].sum()
            if row_sum > target and row_sum > 0:
                p = float(target) / float(row_sum)
                # ensure integer draws
                vals = data[s:e].astype(np.int64, copy=False)
                data[s:e] = np.random.binomial(vals, p).astype(vals.dtype, copy=False)
        X.eliminate_zeros()
        return X
    else:
        X = np.array(X, copy=True)
        # if any non-integers slipped in, round safely
        if not np.issubdtype(X.dtype, np.integer):
            X = np.rint(X).astype(np.int64)
        for i in range(X.shape[0]):
            tot = X[i].sum()
            if tot > target and tot > 0:
                p = float(target) / float(tot)
                X[i] = np.random.binomial(X[i].astype(np.int64, copy=False), p)
        return X

def matched_depth_counts(X_y, X_o, platform):
    """
    Thin both age groups within a platform/stratum to a common target depth
    (min of medians). Returns thinned matrices and the target.
    """
    dy = per_cell_depth(X_y); do = per_cell_depth(X_o)
    target = int(max(1, min(np.median(dy), np.median(do))))
    Xy_ds = downsample_counts_matrix_binomial(X_y, target)
    Xo_ds = downsample_counts_matrix_binomial(X_o, target)
    return Xy_ds, Xo_ds, target
```

```python
def replication_scatter(tissue, supertype, which="a", trim_q=0.99):
    import matplotlib.pyplot as plt
    df = pd.read_csv(OUTDIR/"tables"/f"kinetics_{tissue}_{supertype}.csv")
    a10 = df[df.platform=="10x"][["gene",f"log2FC_{which}"]].rename(columns={f"log2FC_{which}":"x"})
    aSS = df[df.platform=="Smart-seq2"][["gene",f"log2FC_{which}"]].rename(columns={f"log2FC_{which}":"y"})
    m = a10.merge(aSS, on="gene").dropna()
    if len(m)<50:
        print("Not enough overlap to plot.")
        return
    # trim extremes for aesthetics
    xcut = m["x"].abs().quantile(trim_q)
    ycut = m["y"].abs().quantile(trim_q)
    mm = m[(m["x"].abs()<=xcut) & (m["y"].abs()<=ycut)]
    rho = mm[["x","y"]].corr(method="spearman").iloc[0,1]
    agree = (np.sign(mm.x)==np.sign(mm.y)).mean()
    fig, ax = plt.subplots(figsize=(4.5,4.5))
    ax.scatter(mm.x, mm.y, s=4, alpha=0.3)
    ax.axhline(0, lw=0.5, color="k"); ax.axvline(0, lw=0.5, color="k")
    ax.set_xlabel("10x log2FC"); ax.set_ylabel("SS2 log2FC")
    ax.set_title(f"{tissue} — {supertype}\nSpearman ρ={rho:.2f}, sign={agree:.2f}")
    fig.tight_layout()
    fig.savefig(OUTDIR/"figures"/f"replicate_{which}_{tissue}_{supertype}.png", dpi=150)
    plt.close(fig)

# example
replication_scatter("Spleen","B cell","a")
replication_scatter("Spleen","B cell","b")
```

```python
import numpy as np
import pandas as pd

def empirical_null_cut(df, param="a", B=50, seed=1):
    """
    Return a per-stratum, per-parameter effect-size cut:
    the median (over B shuffles) of the 95th percentile of |log2FC_param|
    among genes whose bootstrap CI excludes zero.
    """
    rng = np.random.default_rng(seed)
    eff = df[f"log2FC_{param}"].to_numpy()
    lo  = df[f"boot_lo_{param}"].to_numpy()
    hi  = df[f"boot_hi_{param}"].to_numpy()
    ci_ex = ((lo>0) & (hi>0)) | ((lo<0) & (hi<0))

    vals = []
    for b in range(B):
        # sign-flip null (keeps magnitude distribution; breaks direction)
        sgn = rng.choice([-1, 1], size=eff.shape[0])
        null_eff = eff * sgn
        # 95th percentile of |effect| among CI-passing genes
        if ci_ex.any():
            vals.append(np.quantile(np.abs(null_eff[ci_ex]), 0.95))
    return float(np.nanmedian(vals)) if len(vals) else 0.25
```

```python
import numpy as np, pandas as pd
from pathlib import Path

OUTDIR = Path("/Users/david-johnlonge/Documents/Uni stuff/diss/8273102/results")

def empirical_null_from_nonsig(df, param="a", q=0.95):
    lo = df[f"boot_lo_{param}"].to_numpy()
    hi = df[f"boot_hi_{param}"].to_numpy()
    eff = df[f"log2FC_{param}"].to_numpy()
    # non-significant = CI crosses 0
    nonsig = ~(((lo>0)&(hi>0)) | ((lo<0)&(hi<0)))
    if nonsig.sum() < 100:  # fallback if almost everything looks significant
        return 0.25
    return float(np.nanquantile(np.abs(eff[nonsig]), q))

# Compute per-stratum SS2 cuts from the non-significant set
rows = []
for p in (OUTDIR/"tables").glob("kinetics_*_*.csv"):
    df = pd.read_csv(p)
    ss = df[df.platform=="Smart-seq2"].copy()
    if len(ss) < 200: 
        continue
    t, c = ss["tissue"].iat[0], ss["cell_supertype"].iat[0]
    cutA = max(0.25, empirical_null_from_nonsig(ss, "a", q=0.95))
    cutB = max(0.25, empirical_null_from_nonsig(ss, "b", q=0.95))
    # optionally cap the cut to avoid over-stringency from odd strata
    cutA = float(min(cutA, 1.5))
    cutB = float(min(cutB, 1.5))
    rows.append({"tissue":t,"cell_supertype":c,"platform":"Smart-seq2",
                 "cut_a":cutA, "cut_b":cutB})
cuts_tbl = pd.DataFrame(rows).sort_values(["tissue","cell_supertype"])
cuts_tbl.to_csv(OUTDIR/"tables/ss2_empirical_null_cuts.csv", index=False)
print("Saved:", OUTDIR/"tables/ss2_empirical_null_cuts.csv")
cuts_tbl.head(10)
```

```python
import numpy as np, pandas as pd
from pathlib import Path

ROOT   = Path("/Users/david-johnlonge/Documents/Uni stuff/diss/8273102")
OUTDIR = ROOT/"results"
TABDIR = OUTDIR/"tables"

# load config and strata
CONFIG = json.loads((TABDIR/"CONFIG.json").read_text())
ABS_MIN     = CONFIG["thresholds"]["abs_log2fc_min"]
SS2_CAP     = CONFIG["thresholds"]["ss2_abs_log2fc_cap"]
NULL_Q      = CONFIG["thresholds"]["ss2_null_quantile"]

def empirical_null_from_nonsig(df, param="a", q=NULL_Q):
    lo = df[f"boot_lo_{param}"].to_numpy()
    hi = df[f"boot_hi_{param}"].to_numpy()
    ef = df[f"log2FC_{param}"].to_numpy()
    nonsig = ~(((lo>0)&(hi>0)) | ((lo<0)&(hi<0)))
    if nonsig.sum() < 100:
        return ABS_MIN
    cut = float(np.nanquantile(np.abs(ef[nonsig]), q))
    return float(min(max(cut, ABS_MIN), SS2_CAP))

def strict_mask(df, param, abs_cut):
    lo = df[f"boot_lo_{param}"]; hi = df[f"boot_hi_{param}"]
    ci = ((lo>0)&(hi>0)) | ((lo<0)&(hi<0))
    big = df[f"log2FC_{param}"].abs() >= abs_cut
    return (ci & big).to_numpy()

# compute SS2 cuts and save
cut_rows = []
for f in TABDIR.glob("kinetics_*_*.csv"):
    df = pd.read_csv(f)
    ss = df[df.platform=="Smart-seq2"].copy()
    if ss.empty or len(ss)<200: 
        continue
    t = ss["tissue"].iat[0]; c = ss["cell_supertype"].iat[0]
    cut_a = empirical_null_from_nonsig(ss,"a")
    cut_b = empirical_null_from_nonsig(ss,"b")
    cut_rows.append({"tissue":t,"cell_supertype":c,"platform":"Smart-seq2",
                     "cut_a":cut_a, "cut_b":cut_b})
cuts = (pd.DataFrame(cut_rows)
          .sort_values(["tissue","cell_supertype"])
          .reset_index(drop=True))
cuts.to_csv(TABDIR/"ss2_empirical_null_cuts.csv", index=False)
print("saved:", TABDIR/"ss2_empirical_null_cuts.csv")
display(cuts.head(10))

# map for quick lookup
CUT = {(r.tissue,r.cell_supertype,"a"):r.cut_a for _,r in cuts.iterrows()}
CUT.update({(r.tissue,r.cell_supertype,"b"):r.cut_b for _,r in cuts.iterrows()})

# strict per-stratum summary
sum_rows=[]
for f in TABDIR.glob("kinetics_*_*.csv"):
    df = pd.read_csv(f)
    t = df["tissue"].iat[0]; c = df["cell_supertype"].iat[0]
    for plat in ["10x","Smart-seq2"]:
        sub = df[df.platform==plat].copy()
        if sub.empty: 
            continue
        ca = ABS_MIN if plat=="10x" else CUT.get((t,c,"a"), ABS_MIN)
        cb = ABS_MIN if plat=="10x" else CUT.get((t,c,"b"), ABS_MIN)
        msa = strict_mask(sub,"a", ca)
        msb = strict_mask(sub,"b", cb)
        sum_rows.append({
            "tissue":t,"cell_supertype":c,"platform":plat,
            "n_genes":len(sub),
            "n_sig_a_strict": int(msa.sum()),
            "n_sig_b_strict": int(msb.sum()),
            "frac_sig_a_strict": float(msa.mean()),
            "frac_sig_b_strict": float(msb.mean()),
            "median_log2FC_a": float(sub["log2FC_a"].median(skipna=True)),
            "median_log2FC_b": float(sub["log2FC_b"].median(skipna=True)),
            "abs_cut_a": ca, "abs_cut_b": cb
        })
summary = (pd.DataFrame(sum_rows)
             .sort_values(["tissue","cell_supertype","platform"])
             .reset_index(drop=True))
summary.to_csv(TABDIR/"per_stratum_summary_strict.csv", index=False)
print("saved:", TABDIR/"per_stratum_summary_strict.csv")
display(summary.head(12))

# platform-concordant gene lists per stratum and param
LISTDIR = TABDIR/"gene_lists"
LISTDIR.mkdir(parents=True, exist_ok=True)

conc_rows=[]
for f in TABDIR.glob("kinetics_*_*.csv"):
    df = pd.read_csv(f)
    t = df["tissue"].iat[0]; c = df["cell_supertype"].iat[0]
    d10 = df[df.platform=="10x"].copy()
    dS  = df[df.platform=="Smart-seq2"].copy()
    if d10.empty or dS.empty:
        continue
    ca = CUT.get((t,c,"a"), ABS_MIN); cb = CUT.get((t,c,"b"), ABS_MIN)

    m10a = strict_mask(d10,"a", ABS_MIN); m10b = strict_mask(d10,"b", ABS_MIN)
    mSSa = strict_mask(dS, "a", ca);      mSSb = strict_mask(dS, "b", cb)

    A = d10.loc[m10a,["gene","log2FC_a"]].merge(
        dS.loc[mSSa,["gene","log2FC_a"]], on="gene", suffixes=("_10x","_SS2"))
    B = d10.loc[m10b,["gene","log2FC_b"]].merge(
        dS.loc[mSSb,["gene","log2FC_b"]], on="gene", suffixes=("_10x","_SS2"))

    # same-sign intersection
    A_conc = A[np.sign(A["log2FC_a_10x"])==np.sign(A["log2FC_a_SS2"])].copy()
    B_conc = B[np.sign(B["log2FC_b_10x"])==np.sign(B["log2FC_b_SS2"])].copy()

    # save lists sorted by mean absolute effect
    A_conc.assign(mean_abs=lambda x: (x["log2FC_a_10x"].abs()+x["log2FC_a_SS2"].abs())/2) \
          .sort_values("mean_abs", ascending=False)["gene"] \
          .to_csv(LISTDIR/f"{t}__{c}__concordant_a_genes.txt", index=False, header=False)

    B_conc.assign(mean_abs=lambda x: (x["log2FC_b_10x"].abs()+x["log2FC_b_SS2"].abs())/2) \
          .sort_values("mean_abs", ascending=False)["gene"] \
          .to_csv(LISTDIR/f"{t}__{c}__concordant_b_genes.txt", index=False, header=False)

    conc_rows.append({
        "tissue": t, "cell_supertype": c,
        "n_overlap_a": int(len(A)), "n_concordant_a": int(len(A_conc)),
        "n_overlap_b": int(len(B)), "n_concordant_b": int(len(B_conc))
    })

concord = (pd.DataFrame(conc_rows)
             .sort_values(["n_concordant_a","n_concordant_b"], ascending=False)
             .reset_index(drop=True))
concord.to_csv(TABDIR/"platform_concordant_hits.csv", index=False)
print("saved:", TABDIR/"platform_concordant_hits.csv")
display(concord.head(10))
```

```python
def dotplot(sig, lib, outpath, n_terms=15):
    sub = sig[sig.library==lib].copy()
    if sub.empty:
        print(f"No sig terms for {lib}.")
        return
    sub["stratum"] = sub["tissue"] + " | " + sub["cell_supertype"] + " | " + sub["param"]

    # pick frequent/significant terms across strata
    term_counts = (sub[sub.fdr<0.05]["term"].value_counts().head(n_terms).index.tolist())
    sub = sub[sub["term"].isin(term_counts)]
    # -log10(FDR) for color; overlap size for point size
    sub["mlog10"] = -np.log10(sub["fdr"].clip(lower=1e-300))
    sub["k_overlap"] = sub["overlap"].str.split("/").str[0].astype(int)

    # order axes
    strata = sorted(sub["stratum"].unique(), key=lambda s: (s.split(" | ")[0], s.split(" | ")[2], s.split(" | ")[1]))
    terms  = list(term_counts)

    # build matrix-like arrays
    xs = [strata.index(s) for s in sub["stratum"]]
    ys = [terms.index(t)  for t in sub["term"]]
    sizes = (sub["k_overlap"].values)**1.5 * 10.0
    colors= sub["mlog10"].values

    fig, ax = plt.subplots(figsize=(max(10, len(strata)*0.6), max(6, len(terms)*0.4)))
    sc = ax.scatter(xs, ys, s=sizes, c=colors)
    ax.set_xticks(range(len(strata))); ax.set_xticklabels(strata, rotation=90)
    ax.set_yticks(range(len(terms)));  ax.set_yticklabels(terms)
    cbar = fig.colorbar(sc, ax=ax, label="-log10(FDR)")
    ax.set_title(f"{lib}: top {n_terms} enriched terms by frequency")
    fig.subplots_adjust(left=0.35, bottom=0.45, right=0.98, top=0.90)
    fig.savefig(outpath, dpi=180)
    plt.close(fig)

dotplot(sig, "MSigDB_Hallmark_2020", FIG/"hallmark_dotplot.png", n_terms=12)
dotplot(sig, "Reactome_2022",         FIG/"reactome_dotplot.png",  n_terms=20)
dotplot(sig, "GO_Biological_Process_2023", FIG/"go_bp_dotplot.png", n_terms=20)
```

```python
# --- FIX PACK: stable helpers for the bursting analysis ---

from pathlib import Path
import numpy as np
import pandas as pd
from scipy import sparse

# 0) tiny utils
def _as_csr_counts(adata):
    """Return raw counts (CSR) and gene_names. Prefer layers['counts'] if present."""
    if "counts" in adata.layers:
        X = adata.layers["counts"]
    else:
        X = adata.X
    X = X.tocsr() if sparse.issparse(X) else sparse.csr_matrix(X)
    # prefer gene_symbol if present and unique, else var_names
    if "gene_symbol" in adata.var and adata.var["gene_symbol"].notna().any():
        g = adata.var["gene_symbol"].astype(str).values
    else:
        g = adata.var_names.astype(str).values
    return X, g

def ensure_qc_columns(adata):
    """Make sure obs has n_genes and n_counts for basic QC filters."""
    if "n_counts" not in adata.obs:
        # sum over raw counts
        X, _ = _as_csr_counts(adata)
        adata.obs["n_counts"] = np.asarray(X.sum(axis=1)).ravel()
    if "n_genes" not in adata.obs:
        if "n_genes_by_counts" in adata.obs:
            adata.obs["n_genes"] = adata.obs["n_genes_by_counts"]
        else:
            X, _ = _as_csr_counts(adata)
            adata.obs["n_genes"] = np.diff(X.indptr)

def qc_filter_obs(adata, min_genes=200, max_mito=1.0):
    """Return boolean mask for obs QC (simple and conservative)."""
    ensure_qc_columns(adata)
    ok = adata.obs["n_genes"] >= min_genes
    # optionally add mito % filter if you computed it; default pass-through
    return ok.values

# 1) selection + counts getter (no globals; returns counts, obs, genes)
def get_counts(adata, tissue, cell_supertype, age_group, use_tissue_std=True):
    """Subset adata and return (X_counts, obs_subset, gene_names)."""
    tcol = "tissue_std" if use_tissue_std and "tissue_std" in adata.obs else "tissue"
    if "cell_supertype" not in adata.obs:
        raise ValueError("adata.obs lacks 'cell_supertype' column.")
    idx = (
        (adata.obs[tcol] == tissue) &
        (adata.obs["cell_supertype"] == cell_supertype) &
        (adata.obs["age_group"] == age_group)
    )
    sub = adata[idx].copy()
    # apply minimal QC
    ok = qc_filter_obs(sub)
    sub = sub[ok].copy()
    X, genes = _as_csr_counts(sub)
    return X, sub.obs.copy(), genes

# 2) gene set intersections (within and across matrices)
def intersect_gene_sets(glist, gene_lists):
    """Return intersection gene names and list of index arrays matching each gene_list."""
    inter = set(glist[0])
    for g in gene_lists[1:]:
        inter &= set(g)
    inter = np.array(sorted(inter))
    idxs = [pd.Index(g).get_indexer(inter) for g in gene_lists]
    return inter, idxs

def slice_by_genes(X, idx):
    """Slice a CSR matrix by gene indices (columns)."""
    return X[:, idx] if isinstance(idx, (list, np.ndarray)) else X

# 3) depth matching (binomial thinning; robust for CSR/dense)
def downsample_counts_matrix(X, target):
    """Downsample each cell to ~target UMIs using per-cell binomial thinning."""
    if sparse.issparse(X):
        X = X.tocsr(copy=True)
        out = X.copy()
        indptr = out.indptr
        data = out.data
        # per-cell totals
        totals = np.asarray(X.sum(axis=1)).ravel()
        for i in range(X.shape[0]):
            s, e = indptr[i], indptr[i+1]
            if e == s: 
                continue
            tot = totals[i]
            if tot <= 0 or tot <= target:
                continue
            p = float(target) / float(tot)
            # binomial thin each nonzero entry
            data[s:e] = np.random.binomial(data[s:e].astype(np.int64), p)
        out.eliminate_zeros()
        return out
    else:
        # dense fallback
        out = X.copy().astype(np.int64)
        totals = out.sum(axis=1)
        for i in range(out.shape[0]):
            tot = totals[i]
            if tot <= 0 or tot <= target:
                continue
            p = float(target) / float(tot)
            out[i, :] = np.random.binomial(out[i, :], p)
        return sparse.csr_matrix(out)

def matched_depth_counts(X_y, X_o):
    """Downsample young/old to the min of their median depths."""
    tot_y = np.asarray(X_y.sum(axis=1)).ravel()
    tot_o = np.asarray(X_o.sum(axis=1)).ravel()
    target = int(max(1, min(np.median(tot_y), np.median(tot_o))))
    return downsample_counts_matrix(X_y, target), downsample_counts_matrix(X_o, target), target

# 4) gene filters (apply AFTER intersections)
def gene_filter_both(Xy, Xo, min_cells_frac=0.10, min_mean=0.01):
    """Keep genes expressed in >= min_cells_frac of cells in BOTH ages and mean >= min_mean."""
    def mask_one(X):
        n = X.shape[0]
        cells = np.asarray((X>0).sum(axis=0)).ravel() if sparse.issparse(X) else (X>0).sum(axis=0)
        mean  = np.asarray(X.mean(axis=0)).ravel()
        return (cells >= n * min_cells_frac) & (mean >= min_mean)
    my = mask_one(Xy)
    mo = mask_one(Xo)
    return my & mo

# 5) MoM + bootstrap (no globals; robust to zeros; returns tidy df)
def _mom_nb(y):
    """Method-of-moments NB burst proxies (â, b̂) with guards."""
    mu = y.mean()
    v  = y.var(ddof=1)
    if not np.isfinite(mu) or mu <= 0:
        return np.nan, np.nan
    # NB: Var = mu + mu^2 / a  => a = mu^2 / (Var - mu)
    denom = v - mu
    a = np.nan if denom <= 0 else (mu*mu)/denom
    # b̂ = (Var - mu)/mu
    b = np.nan if denom <= 0 else denom/mu
    return a, b

def per_gene_bootstrap(X_o, X_y, genes, n_boot=200, seed=1):
    """Bootstrap cells to get medians/CI for â and b̂, then log2FC between old and young."""
    rng = np.random.default_rng(seed)
    # precompute per-gene estimates
    def est_batch(X):
        # X: cells x genes CSR
        res_a = []; res_b = []
        # iterate columns efficiently
        for g in range(X.shape[1]):
            col = X[:, g].toarray().ravel() if sparse.issparse(X) else X[:, g]
            a, b = _mom_nb(col)
            res_a.append(a); res_b.append(b)
        return np.array(res_a), np.array(res_b)

    a_o, b_o = est_batch(X_o)
    a_y, b_y = est_batch(X_y)

    # bootstrapped medians
    def boot_stats(X):
        n = X.shape[0]
        A = []; B = []
        for _ in range(n_boot):
            idx = rng.integers(0, n, size=n)
            Xb = X[idx]
            a, b = est_batch(Xb)
            A.append(a); B.append(b)
        A = np.vstack(A); B = np.vstack(B)
        med_a = np.nanmedian(A, axis=0)
        lo_a  = np.nanpercentile(A, 2.5, axis=0); hi_a = np.nanpercentile(A, 97.5, axis=0)
        med_b = np.nanmedian(B, axis=0)
        lo_b  = np.nanpercentile(B, 2.5, axis=0); hi_b = np.nanpercentile(B, 97.5, axis=0)
        return med_a, lo_a, hi_a, med_b, lo_b, hi_b

    med_ao, lo_ao, hi_ao, med_bo, lo_bo, hi_bo = boot_stats(X_o)
    med_ay, lo_ay, hi_ay, med_by, lo_by, hi_by = boot_stats(X_y)

    # log2FC using medians; guard zeros
    def log2fc(old, yng, eps=1e-8):
        return np.log2((old+eps)/(yng+eps))

    df = pd.DataFrame({
        "gene": genes,
        "a_old": med_ao, "a_yng": med_ay,
        "b_old": med_bo, "b_yng": med_by,
        "a_old_lo": lo_ao, "a_old_hi": hi_ao,
        "a_yng_lo": lo_ay, "a_yng_hi": hi_ay,
        "b_old_lo": lo_bo, "b_old_hi": hi_bo,
        "b_yng_lo": lo_by, "b_yng_hi": hi_by,
        "log2FC_a": log2fc(med_ao, med_ay),
        "log2FC_b": log2fc(med_bo, med_by),
    })
    # crude p-values from bootstrap sign tests
    def boot_p(x_old, x_yng):
        # two-sided fraction of boot medians with opposite sign or smaller magnitude
        return np.full_like(x_old, np.nan, dtype=float)  # keep simple; you already do empirical cutoffs later
    df["p_a"] = np.nan; df["p_b"] = np.nan
    return df

# 6) empirical null cutoffs (fix .quantile on ndarray)
def empirical_null_cut(df, param="a", B=100, seed=1):
    """Estimate |log2FC| 95th percentile by random sign flips (within platform/stratum)."""
    rng = np.random.default_rng(seed)
    eff = df[[f"log2FC_{param}"]].values.ravel()
    ci  = pd.to_numeric(df[f"q_{param}"], errors="coerce").values if f"q_{param}" in df else None
    vals = []
    for _ in range(B):
        sgn = rng.choice([-1,1], size=eff.size)
        null_eff = eff * sgn
        vals.append(np.nanquantile(np.abs(null_eff), 0.95))
    return float(np.nanmedian(vals))
```

```python
from scipy import sparse

# --- additional gene filter: require NB-like overdispersion in both ages
def overdispersed_both(Xy, Xo):
    def od(X):
        # sample var > sample mean (per gene)
        if sparse.issparse(X):
            mu = np.asarray(X.mean(axis=0)).ravel()
            # var by rows: cheap approx; for sparse we’ll compute column-wise using toarray() in chunks
            V = []
            B = 1024
            for k in range(0, X.shape[1], B):
                block = X[:, k:k+B].toarray()
                V.append(block.var(axis=0, ddof=1))
            var = np.concatenate(V)
        else:
            mu  = X.mean(axis=0)
            var = X.var(axis=0, ddof=1)
        return var > mu
    return od(Xy) & od(Xo)

def run_stratum(ad10x, adss2, tissue, supertype, min_cells_frac=0.10, min_mean=0.01, n_boot=200, seed=123):
    # 10x: young/old
    Xy10, obs_y10, g_y10 = get_counts(ad10x, tissue, supertype, "young")
    Xo10, obs_o10, g_o10 = get_counts(ad10x, tissue, supertype, "old")
    if Xy10.shape[0] < 50 or Xo10.shape[0] < 50: 
        return None

    # intersect within platform
    genes10, [iy10, io10] = intersect_gene_sets([g_y10, g_o10], [g_y10, g_o10])
    Xy10 = slice_by_genes(Xy10, iy10); Xo10 = slice_by_genes(Xo10, io10)

    # depth-match
    Xy10, Xo10, tgt10 = matched_depth_counts(Xy10, Xo10)

    # basic filters
    mask10 = gene_filter_both(Xy10, Xo10, min_cells_frac, min_mean) & overdispersed_both(Xy10, Xo10)
    if mask10.sum() < 200: 
        return None
    Xy10 = Xy10[:, mask10]; Xo10 = Xo10[:, mask10]
    genes10 = genes10[mask10]

    res10 = per_gene_bootstrap(Xo10, Xy10, genes10, n_boot=n_boot, seed=seed)
    res10["platform"] = "10x"
    res10["tissue"] = tissue
    res10["cell_supertype"] = supertype

    # SS2: young/old (same steps)
    XyS, obs_yS, g_yS = get_counts(adss2, tissue, supertype, "young")
    XoS, obs_oS, g_oS = get_counts(adss2, tissue, supertype, "old")
    if XyS.shape[0] < 50 or XoS.shape[0] < 50:
        return res10

    genesS, [iyS, ioS] = intersect_gene_sets([g_yS, g_oS], [g_yS, g_oS])
    XyS = slice_by_genes(XyS, iyS); XoS = slice_by_genes(XoS, ioS)
    XyS, XoS, tgtS = matched_depth_counts(XyS, XoS)
    maskS = gene_filter_both(XyS, XoS, min_cells_frac, min_mean) & overdispersed_both(XyS, XoS)
    if maskS.sum() >= 200:
        XyS = XyS[:, maskS]; XoS = XoS[:, maskS]
        genesS = genesS[maskS]
        resS = per_gene_bootstrap(XoS, XyS, genesS, n_boot=n_boot, seed=seed+1)
        resS["platform"] = "Smart-seq2"
        resS["tissue"] = tissue
        resS["cell_supertype"] = supertype
        return pd.concat([res10, resS], ignore_index=True)
    else:
        return res10
```

```python
import numpy as np
import pandas as pd

def _effect_col(df: pd.DataFrame, param: str) -> str:
    for c in (f"log2FC_{param}", f"log2fc_{param}", f"logFC_{param}", f"logfc_{param}"):
        if c in df.columns: 
            return c
    raise KeyError(f"Need log2FC column for param={param}; got {df.columns.tolist()}")

def empirical_null_cut(df: pd.DataFrame, param: str, B: int = 100, seed: int = 1) -> float:
    """95th pct of |effect| under random sign-flip; ignores NaN/inf; returns NaN if no data."""
    col = _effect_col(df, param)
    eff = df[col].to_numpy(dtype=float)
    eff = eff[np.isfinite(eff)]                 # drop NaN/inf
    if eff.size < 50:                           # too few genes -> unstable null
        return np.nan
    rng = np.random.default_rng(seed)
    qs = []
    for _ in range(B):
        sgn = rng.choice([-1.0, 1.0], size=eff.size)
        qs.append(np.quantile(np.abs(eff * sgn), 0.95))
    return float(np.nanmedian(qs))
```

```python
def platform_concordance(df, param):
    col = f"strict_{param}"
    lfc = f"log2FC_{param}"
    out = []
    for (t,c), sub in df.groupby(["tissue","cell_supertype"]):
        a = sub[sub.platform=="10x"][["gene", col, lfc]].rename(columns={col:"hit10", lfc:"fc10"})
        b = sub[sub.platform=="Smart-seq2"][["gene", col, lfc]].rename(columns={col:"hitSS", lfc:"fcSS"})
        m = a.merge(b, on="gene", how="inner")
        if m.empty: continue
        # concordant = both pass + same sign
        m["same_sign"] = np.sign(m["fc10"]) == np.sign(m["fcSS"])
        concord = m["hit10"] & m["hitSS"] & m["same_sign"]
        out.append({
            "tissue":t, "cell_supertype":c,
            "param":param,
            "n_overlap": int(len(m)),
            "n_concordant": int(concord.sum()),
            "frac_concordant": float(concord.mean()),
        })
    return pd.DataFrame(out)

ALLs = pd.read_csv(OUTDIR/"tables"/"all_strata_kinetics_with_strict.csv")
rep_a = platform_concordance(ALLs, "a")
rep_b = platform_concordance(ALLs, "b")
rep = pd.concat([rep_a, rep_b], ignore_index=True)
rep.to_csv(OUTDIR/"tables"/"cross_platform_concordance_strict.csv", index=False)
print("Saved:", OUTDIR/"tables"/"cross_platform_concordance_strict.csv")
```

```python
import re, numpy as np, pandas as pd

# 1) Normalize column names
GSEA.columns = (GSEA.columns.str.strip()
                           .str.lower()
                           .str.replace(" ", "_")
                           .str.replace("-", "_"))

print("GSEA cols:", list(GSEA.columns))

# 2) Find column aliases
def pick(cols, options):
    for name in options:
        if name in cols: return name
    # try fuzzy contains
    for c in cols:
        if any(re.fullmatch(pat, c) for pat in options): return c
    return None

TERM_COL = pick(GSEA.columns, ["term","name","pathway"])
NES_COL  = pick(GSEA.columns, ["nes","normalized_enrichment_score"])
P_COL    = pick(GSEA.columns, ["pval","p_value","pvalue"])
FDR_COL  = pick(GSEA.columns, ["fdr","fdr_q","fdr_q_val","padj","adj_p","qval","q_value","fdr_bh"])

# 3) If FDR missing, compute BH-FDR from p-values (per stratum to be safest)
def bh_fdr(p):
    p = np.asarray(p, float)
    n = p.size
    order = np.argsort(p)
    ranked = p[order]
    q = ranked * n / (np.arange(1, n+1))
    q = np.minimum.accumulate(q[::-1])[::-1]
    out = np.empty_like(q)
    out[order] = q
    return np.clip(out, 0, 1)

if FDR_COL is None:
    assert P_COL is not None, "No p-value column found to compute FDR."
    GSEA["fdr"] = np.nan
    grp = ["tissue","cell_supertype","param","platform","library"]
    for _, idx in GSEA.groupby(grp).groups.items():
        pv = GSEA.loc[idx, P_COL].astype(float).values
        GSEA.loc[idx, "fdr"] = bh_fdr(pv)
    FDR_COL = "fdr"

# 4) Safety checks
for need, name in [(TERM_COL,"TERM_COL"), (NES_COL,"NES_COL"), (FDR_COL,"FDR_COL")]:
    if need is None: raise ValueError(f"Missing a required column; check mapping for {name}.")
```

```python
# Composite ranking of strata to pick ~5 "flagship" systems
# Reads your existing outputs and computes a transparent score per (tissue × cell_supertype).

import pandas as pd
import numpy as np
from pathlib import Path
from caas_jupyter_tools import display_dataframe_to_user

ROOT = Path("/Users/david-johnlonge/Documents/Uni stuff/diss/8273102")
TABS = ROOT / "results" / "tables"

# ---- Load inputs (be tolerant to file presence) ----
per_strict = pd.read_csv(TABS / "per_stratum_summary_strict.csv")
rep_genes  = pd.read_csv(TABS / "cross_platform_replication.csv") if (TABS / "cross_platform_replication.csv").exists() else pd.DataFrame()
path_rep   = pd.read_csv(TABS / "enrichment" / "pathway_cross_platform_replication.csv") if (TABS / "enrichment" / "pathway_cross_platform_replication.csv").exists() else pd.DataFrame()
cons_counts = pd.read_csv(TABS / "enrichment" / "consensus_gene_counts.csv") if (TABS / "enrichment" / "consensus_gene_counts.csv").exists() else pd.DataFrame()

# ---- 1) Strict-hit strength (average across platforms & params) ----
# Use mean of frac_sig_a_strict and frac_sig_b_strict per stratum×platform, then average across platforms.
if not per_strict.empty:
    tmp = per_strict.assign(frac_mean=lambda d: (d["frac_sig_a_strict"] + d["frac_sig_b_strict"]) / 2)
    strict_strength = (tmp.groupby(["tissue","cell_supertype"], as_index=False)
                         .agg(strict_frac=("frac_mean","mean"),
                              n_genes=("n_genes","sum")))
else:
    strict_strength = pd.DataFrame(columns=["tissue","cell_supertype","strict_frac","n_genes"])

# ---- 2) Cross-platform gene-level replication (rho/sign) ----
if not rep_genes.empty:
    rep_gene_strength = (rep_genes
                         .assign(rho_abs=lambda d: (d["rho_a"].abs() + d["rho_b"].abs())/2,
                                 sign_mean=lambda d: (d["sign_a"] + d["sign_b"]) / 2)
                         [["tissue","cell_supertype","n_overlap","rho_abs","sign_mean"]])
else:
    rep_gene_strength = pd.DataFrame(columns=["tissue","cell_supertype","n_overlap","rho_abs","sign_mean"])

# ---- 3) Pathway overlap replication (max Jaccard across param/direction) ----
if not path_rep.empty:
    # file columns: tissue, cell, param, direction, jaccard_terms, n10, nSS2
    path_rep2 = (path_rep
                 .rename(columns={"cell":"cell_supertype"})
                 .groupby(["tissue","cell_supertype"], as_index=False)
                 .agg(pathway_jaccard_max=("jaccard_terms","max"),
                      pathway_jaccard_mean=("jaccard_terms","mean")))
else:
    path_rep2 = pd.DataFrame(columns=["tissue","cell_supertype","pathway_jaccard_max","pathway_jaccard_mean"])

# ---- 4) Consensus gene counts (strict hits that replicate) ----
if not cons_counts.empty:
    # expected columns: tissue, cell_supertype, cons_a, cons_b
    if "cons_a" not in cons_counts.columns or "cons_b" not in cons_counts.columns:
        # try alternative column names
        for a_alias, b_alias in [("consensus_a","consensus_b"), ("n_cons_a","n_cons_b")]:
            if a_alias in cons_counts.columns and b_alias in cons_counts.columns:
                cons_counts = cons_counts.rename(columns={a_alias:"cons_a", b_alias:"cons_b"})
                break
    cons_counts["cons_total"] = cons_counts.get("cons_a", 0) + cons_counts.get("cons_b", 0)
    cons_strength = cons_counts[["tissue","cell_supertype","cons_total"]]
else:
    cons_strength = pd.DataFrame(columns=["tissue","cell_supertype","cons_total"])

# ---- Join everything ----
dfs = [strict_strength, rep_gene_strength, path_rep2, cons_strength]
from functools import reduce
rankdf = reduce(lambda left,right: pd.merge(left, right, on=["tissue","cell_supertype"], how="outer"), dfs).fillna(0)

# ---- Scale metrics to 0-1 (robust) ----
def minmax(s: pd.Series):
    s = s.astype(float)
    lo, hi = s.min(), s.max()
    return (s - lo) / (hi - lo) if hi > lo else pd.Series(np.zeros(len(s)), index=s.index)

rankdf["m_strict"]   = minmax(rankdf.get("strict_frac", pd.Series(np.zeros(len(rankdf)))))
rankdf["m_rho"]      = minmax(rankdf.get("rho_abs", pd.Series(np.zeros(len(rankdf)))))
rankdf["m_sign"]     = rankdf.get("sign_mean", 0.0) # already 0-1
rankdf["m_path"]     = rankdf.get("pathway_jaccard_max", 0.0) # already 0-1
rankdf["m_cons"]     = minmax(rankdf.get("cons_total", pd.Series(np.zeros(len(rankdf)))))

# Optional: weight overlap size a bit (more n_overlap -> more trust in rho/sign)
if "n_overlap" in rankdf.columns and rankdf["n_overlap"].max() > 0:
    w_overlap = 0.5 + 0.5*minmax(rankdf["n_overlap"])  # 0.5–1.0 weight
else:
    w_overlap = 1.0

# ---- Composite score (weights you can tweak) ----
w = dict(strict=0.25, gene_rep=0.25, path=0.30, cons=0.20)  # sums to 1
rankdf["gene_rep"] = (rankdf["m_rho"]*0.4 + rankdf["m_sign"]*0.6) * w_overlap

rankdf["score"] = (
    w["strict"]   * rankdf["m_strict"] +
    w["gene_rep"] * rankdf["gene_rep"] +
    w["path"]     * rankdf["m_path"] +
    w["cons"]     * rankdf["m_cons"]
)

out_cols = [
    "tissue","cell_supertype",
    "score",
    "strict_frac","cons_total","rho_abs","sign_mean","n_overlap",
    "pathway_jaccard_max",
]
ranked = (rankdf[out_cols]
          .sort_values("score", ascending=False)
          .reset_index(drop=True))

# Save + show
out_path = TABS / "flagship_strata_ranking.csv"
ranked.to_csv(out_path, index=False)
display_dataframe_to_user("Flagship strata ranking (higher is better)", ranked)

out_path.as_posix()
```

```python
import pandas as pd, numpy as np
from pathlib import Path

ROOT = Path("/Users/david-johnlonge/Documents/Uni stuff/diss/8273102")
TABS = ROOT / "results" / "tables"

# --- Load inputs
strict = pd.read_csv(TABS / "per_stratum_summary_strict.csv")  # frac_sig_*_strict by platform
gene_rep = pd.read_csv(TABS / "cross_platform_replication.csv")  # rho/sign/n_overlap
path_rep = pd.read_csv(TABS / "enrichment" / "pathway_cross_platform_replication.csv")  # jaccard_terms
cons_path1 = TABS / "enrichment" / "consensus_gene_counts.csv"
cons_path2 = TABS / "consensus_gene_counts.csv"
cons = pd.read_csv(cons_path1 if cons_path1.exists() else cons_path2)

# --- Aggregate strict fractions across a/b and platforms
strict_agg = (strict
    .assign(frac = (strict["frac_sig_a_strict"] + strict["frac_sig_b_strict"]) / 2)
    .groupby(["tissue","cell_supertype"], as_index=False)
    .agg(strict_frac_mean=("frac","mean"),
         n_genes_total=("n_genes","sum"))
)

# --- Gene-level cross-platform replication -> single score
def _norm_rho(x): return (x + 1) / 2  # [-1,1] -> [0,1]
gr = (gene_rep
    .assign(rho_a_n=lambda d: _norm_rho(d["rho_a"]),
            rho_b_n=lambda d: _norm_rho(d["rho_b"]))
    .groupby(["tissue","cell_supertype"], as_index=False)
    .agg(rho_a_n=("rho_a_n","mean"),
         rho_b_n=("rho_b_n","mean"),
         sign_a=("sign_a","mean"),
         sign_b=("sign_b","mean"),
         n_overlap=("n_overlap","mean"))
)
# emphasize sign agreement, lightly reward overlap size
gr["gene_rep_score"] = 0.4*((gr["rho_a_n"]+gr["rho_b_n"])/2) + 0.6*((gr["sign_a"]+gr["sign_b"])/2)
gr["n_overlap_s"] = (gr["n_overlap"] - gr["n_overlap"].min()) / (gr["n_overlap"].max() - gr["n_overlap"].min() + 1e-9)
gr["gene_rep_score"] = 0.85*gr["gene_rep_score"] + 0.15*gr["n_overlap_s"]

# --- Pathway Jaccard across platforms, averaged over param/direction
pj = (path_rep
    .groupby(["tissue","cell"], as_index=False)
    .agg(path_jaccard_mean=("jaccard_terms","mean"),
         n_pairs=("jaccard_terms","size"))
    .rename(columns={"cell":"cell_supertype"})
)

# --- Consensus genes across platforms
cons["cons_total"] = cons.filter(like="cons_").sum(axis=1)
cg = cons[["tissue","cell_supertype","cons_total"]]

# --- Merge & scale metrics to 0–1
def minmax(s): return (s - s.min()) / (s.max() - s.min() + 1e-9)

m = (strict_agg.merge(gr, on=["tissue","cell_supertype"], how="left")
              .merge(pj, on=["tissue","cell_supertype"], how="left")
              .merge(cg, on=["tissue","cell_supertype"], how="left")
).fillna({"gene_rep_score":0, "path_jaccard_mean":0, "cons_total":0})

m["strict_s"]   = minmax(m["strict_frac_mean"])
m["gene_rep_s"] = minmax(m["gene_rep_score"])
m["path_s"]     = minmax(m["path_jaccard_mean"])
m["cons_s"]     = minmax(m["cons_total"])

# Composite flagship score (tweak weights if you like)
m["flagship_score"] = 0.25*m["strict_s"] + 0.25*m["gene_rep_s"] + 0.30*m["path_s"] + 0.20*m["cons_s"]

ranked = (m.sort_values("flagship_score", ascending=False)
            [["tissue","cell_supertype","strict_frac_mean","gene_rep_score",
              "path_jaccard_mean","cons_total","flagship_score"]])

outf = TABS / "flagship_strata_ranking.csv"
ranked.to_csv(outf, index=False)
print("Saved:", outf)
print(ranked.head(10).to_string(index=False))
```

```python
import pandas as pd, numpy as np
from pathlib import Path
from scipy.stats import spearmanr

ROOT = Path("/Users/david-johnlonge/Documents/Uni stuff/diss/8273102")
TABS = ROOT/"results/tables"
ALL  = pd.read_csv(TABS/"all_strata_kinetics.csv")

def _finite(d, col): 
    return d[np.isfinite(d[col])].copy()

def platform_replication_summary(all_df):
    rows=[]
    for (t, c), block in all_df.groupby(["tissue","cell_supertype"]):
        # per param: take 10x vs Smart-seq2, inner-join on gene
        rec = {"tissue":t, "cell_supertype":c}
        for prm in ["a","b"]:
            col = f"log2FC_{prm}"
            ten = (_finite(block[block.platform=="10x"], col)[["gene", col]]
                      .rename(columns={col:f"{col}_10x"}))
            ss2 = (_finite(block[block.platform=="Smart-seq2"], col)[["gene", col]]
                      .rename(columns={col:f"{col}_ss2"}))
            M = ten.merge(ss2, on="gene", how="inner")
            rec["n_overlap"] = int(M.shape[0]) if "n_overlap" not in rec else max(rec["n_overlap"], int(M.shape[0]))
            if len(M) >= 20:
                r, _ = spearmanr(M[f"{col}_10x"], M[f"{col}_ss2"])
                rec[f"rho_{prm}"] = float(r)
                s10 = np.sign(M[f"{col}_10x"].to_numpy())
                sS2 = np.sign(M[f"{col}_ss2"].to_numpy())
                rec[f"sign_{prm}"] = float(np.mean(s10==sS2))
            else:
                rec[f"rho_{prm}"]  = np.nan
                rec[f"sign_{prm}"] = np.nan
        rows.append(rec)
    return pd.DataFrame(rows)

rep = platform_replication_summary(ALL).sort_values("tissue")
out = TABS/"platform_replication_summary.csv"
rep.to_csv(out, index=False)
print("Saved:", out)
rep.head()
```

```python
# === Cross-platform replication bars (polished) ===
import pandas as pd, numpy as np, matplotlib.pyplot as plt
from pathlib import Path
from scipy.stats import spearmanr

ROOT = Path("/Users/david-johnlonge/Documents/Uni stuff/diss/8273102")
TABS = ROOT/"results"/"tables"
OUT  = ROOT/"results"/"figures"/"panels"
OUT.mkdir(parents=True, exist_ok=True)

# --------- helpers ---------
def clean_label(t, c):
    repl = {
        "skeletal muscle satellite cell": "Skel. satellite",
        "Mononuclear phagocyte": "Mono. phagocyte",
        "Urothelial cell": "Urothelial",
        "Endothelial cell": "Endothelial",
    }
    c2 = repl.get(c, c)
    return f"{t}–{c2}"

def finite(x):
    return np.isfinite(x) & (x.notna())

def pair_metrics(df10, dfS2, eff_col):
    # align by gene
    m = (df10.set_index("gene")[[eff_col]]
               .join(dfS2.set_index("gene")[[eff_col]], how="inner", lsuffix="_10x", rsuffix="_ss2")
               .rename(columns={f"{eff_col}_10x":"tenx", f"{eff_col}_ss2":"ss2"})
         )
    m = m[finite(m["tenx"]) & finite(m["ss2"])]
    n = len(m)
    if n == 0:
        return np.nan, np.nan, 0
    rho = spearmanr(m["tenx"], m["ss2"]).statistic
    nz   = (m["tenx"] != 0) & (m["ss2"] != 0)
    sign = (np.sign(m.loc[nz,"tenx"]) == np.sign(m.loc[nz,"ss2"])).mean() if nz.any() else np.nan
    return float(rho), float(sign), int(n)

# --------- load and compute ---------
ALL = pd.read_csv(TABS/"all_strata_kinetics.csv")  # must contain: gene, log2FC_a, log2FC_b, platform, tissue, cell_supertype

rows = []
for (t, c), block in ALL.groupby(["tissue","cell_supertype"], dropna=False):
    d10 = block[block.platform=="10x"]
    dS2 = block[block.platform=="Smart-seq2"]
    if d10.empty or dS2.empty:
        continue
    for prm in ("a","b"):
        rho, sign, n = pair_metrics(d10, dS2, f"log2FC_{prm}")
        rows.append({"tissue":t, "cell_supertype":c, "param":prm, "rho":rho, "sign":sign, "n":n})

R = pd.DataFrame(rows)
if R.empty:
    raise RuntimeError("No strata with both platforms found.")

# put into wide for plotting
P = (R.pivot(index=["tissue","cell_supertype"], columns="param", values=["rho","sign"])
       .sort_index())
# order by sign(a) descending
order = (P["sign"]["a"].fillna(-1)).sort_values(ascending=False).index.tolist()
P = P.loc[order]

labels = [clean_label(t,c) for (t,c) in P.index]

# --------- style ---------
import matplotlib as mpl
mpl.rcParams.update({
    "figure.dpi": 300, "savefig.dpi": 300,
    "axes.titlesize": 20, "axes.labelsize": 14,
    "xtick.labelsize": 11, "ytick.labelsize": 11,
    "legend.fontsize": 12, "axes.grid": True, "grid.alpha": 0.15,
    "axes.spines.top": False, "axes.spines.right": False,
})

fig, axs = plt.subplots(1, 2, figsize=(14, 5), constrained_layout=True)

# left: Spearman rho
ax = axs[0]
width = 0.38
x = np.arange(len(P))
ax.bar(x - width/2, P["rho"]["a"].values, width, label="a")
ax.bar(x + width/2, P["rho"]["b"].values, width, label="b")
ax.axhline(0, lw=0.8, color="k", alpha=0.4)
ax.set_xticks(x, labels, rotation=40, ha="right")
ax.set_ylim(-0.1, 0.3)     # your rhos are small and can be slightly negative
ax.set_ylabel("Spearman ρ")
ax.set_title("Spearman ρ")
ax.legend(loc="upper left", bbox_to_anchor=(0.0, 1.15), ncol=2, frameon=False)

# right: sign agreement
ax = axs[1]
ax.bar(x - width/2, P["sign"]["a"].values, width, label="a")
ax.bar(x + width/2, P["sign"]["b"].values, width, label="b")
ax.set_xticks(x, labels, rotation=40, ha="right")
ax.set_ylim(0, 1.0)
ax.set_ylabel("Sign agreement (fraction)")
ax.set_title("Sign agreement")
ax.legend(loc="upper left", bbox_to_anchor=(0.0, 1.15), ncol=2, frameon=False)

# save
out = OUT/"cross_platform_replication_bars_v2.png"
fig.savefig(out, bbox_inches="tight")
print("Saved:", out)
```

```python
# ==== Pretty per-stratum panels (2x2: violins for a,b + cross-platform scatters) ====
import numpy as np, pandas as pd, matplotlib.pyplot as plt
from pathlib import Path
from scipy.stats import spearmanr

ROOT = Path("/Users/david-johnlonge/Documents/Uni stuff/diss/8273102")
TABS = ROOT/"results"/"tables"
OUT  = ROOT/"results"/"figures"/"panels"
OUT.mkdir(parents=True, exist_ok=True)

ALL = pd.read_csv(TABS/"all_strata_kinetics.csv")  # needs: a_old,a_yng,b_old,b_yng,log2FC_a,log2FC_b,platform,tissue,cell_supertype,gene

# ---------- helpers ----------
def finite_ser(x): 
    return x.replace([np.inf, -np.inf], np.nan).dropna()

def q_lims(vals, qlo=0.01, qhi=0.99):
    v = finite_ser(pd.Series(np.concatenate([v for v in vals if len(v)>0], dtype=float))) if len(vals) else pd.Series([], dtype=float)
    if v.empty: 
        return (0, 1)
    lo, hi = np.quantile(v, [qlo, qhi])
    # keep floor at 0 if that makes sense
    return (min(0, lo), hi)

def merge_fc(sub10, subS2, param):
    # align 10x vs SS2 log2FC for scatter
    a = sub10.set_index("gene")[f"log2FC_{param}"]
    b = subS2.set_index("gene")[f"log2FC_{param}"]
    m = pd.concat([a.rename("tenx"), b.rename("ss2")], axis=1).dropna()
    m = m[np.isfinite(m["tenx"]) & np.isfinite(m["ss2"])]
    return m

def sign_agreement(m):
    nz = (m["tenx"]!=0) & (m["ss2"]!=0)
    return (np.sign(m.loc[nz,"tenx"]) == np.sign(m.loc[nz,"ss2"])).mean() if nz.any() else np.nan

param_label = {"a": r"$\hat{\alpha}$", "b": r"$\hat{b}$"}

def make_panel(tissue, cell):
    sub = ALL.query("tissue==@tissue & cell_supertype==@cell")
    ten = sub[sub.platform=="10x"].copy()
    ss2 = sub[sub.platform=="Smart-seq2"].copy()
    if ten.empty or ss2.empty:
        print(f"Skip {tissue} — {cell}: missing one platform")
        return

    # violin data (order: 10x old, 10x young, SS2 old, SS2 young)
    V = {
        "a": [finite_ser(ten["a_old"]).values, finite_ser(ten["a_yng"]).values,
              finite_ser(ss2["a_old"]).values, finite_ser(ss2["a_yng"]).values],
        "b": [finite_ser(ten["b_old"]).values, finite_ser(ten["b_yng"]).values,
              finite_ser(ss2["b_old"]).values, finite_ser(ss2["b_yng"]).values],
    }
    xlabels = ["10x old","10x young","SS2 old","SS2 young"]

    # scatters
    M = {k: merge_fc(ten, ss2, k) for k in ("a","b")}
    stats = {}
    for k in ("a","b"):
        if len(M[k])>2:
            rho = spearmanr(M[k]["tenx"], M[k]["ss2"]).statistic
            sign = sign_agreement(M[k])
        else:
            rho, sign = np.nan, np.nan
        stats[k] = (rho, sign, len(M[k]))

    # ---------- plotting ----------
    import matplotlib as mpl
    mpl.rcParams.update({
        "figure.dpi": 300, "savefig.dpi": 300,
        "axes.titlesize": 14, "axes.labelsize": 12,
        "xtick.labelsize": 10, "ytick.labelsize": 10,
        "axes.grid": True, "grid.alpha": 0.15, "axes.spines.top": False, "axes.spines.right": False,
    })

    fig = plt.figure(figsize=(10.5, 7.5), constrained_layout=True)
    gs  = fig.add_gridspec(2, 2, height_ratios=[1, 1.2], wspace=0.25, hspace=0.25)

    # top row: violins
    for j, k in enumerate(("a","b")):
        ax = fig.add_subplot(gs[0, j])
        parts = ax.violinplot(V[k], showmeans=False, showextrema=False, widths=0.8)
        for pc in parts['bodies']:
            pc.set_alpha(0.6)
        ax.set_xticks(np.arange(1,5), xlabels, rotation=20, ha="right")
        ax.set_ylabel(param_label[k])
        lo, hi = q_lims(V[k])   # trim extreme tails
        if hi > lo: ax.set_ylim(lo, hi)
        ax.set_title(f"{tissue} — {cell} ({param_label[k]})")

    # bottom row: scatters
    for j, k in enumerate(("a","b")):
        ax = fig.add_subplot(gs[1, j])
        m = M[k]
        ax.scatter(m["tenx"], m["ss2"], s=8, alpha=0.35)
        lims = np.array([
            np.nanpercentile(m["tenx"], [1, 99]).tolist() if len(m) else [-1,1],
            np.nanpercentile(m["ss2"], [1, 99]).tolist() if len(m) else [-1,1]
        ])
        lo, hi = float(np.min(lims)), float(np.max(lims))
        pad = 0.05*(hi-lo)
        ax.set_xlim(lo-pad, hi+pad); ax.set_ylim(lo-pad, hi+pad)
        ax.axhline(0, lw=0.8, color="k", alpha=0.4); ax.axvline(0, lw=0.8, color="k", alpha=0.4)
        ax.plot([lo-pad, hi+pad], [lo-pad, hi+pad], ls="--", lw=0.8, alpha=0.5)  # y=x
        rho, sgn, n = stats[k]
        ax.set_xlabel("10x log2FC"); ax.set_ylabel("SS2 log2FC")
        ax.set_title(f"10x vs SS2 ({param_label[k]}): ρ={rho:.2f}  sign={sgn:.2f}  n={n}")

    fig.suptitle(f"{tissue} — {cell}", fontsize=16)
    out = OUT/f"panel_{tissue}_{cell}.png"
    fig.savefig(out, bbox_inches="tight")
    plt.close(fig)
    print("Saved panel:", out)

# ---------- run for the six strata (2–7) ----------
todo = [
    ("Tongue", "Keratinocyte"),
    ("Marrow", "Mononuclear phagocyte"),
    ("Bladder", "Urothelial cell"),
    ("Lung", "Smooth muscle"),
    ("Marrow", "Granulocyte"),
    ("Liver", "Hepatocyte"),
]
for t, c in todo:
    make_panel(t, c)
```

```python
# =========================
# FLAGSHIP PATHWAY FIGURES: heatmaps + 4-panel bar charts
# =========================
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
import textwrap

ROOT = Path("/Users/david-johnlonge/Documents/Uni stuff/diss/8273102")
TABS = ROOT/"results"/"tables"
FIGS = ROOT/"results"/"figures"/"panels"
FIGS.mkdir(parents=True, exist_ok=True)

# --- inputs
E = pd.read_csv(TABS/"enrichment/enrichr_all.csv")          # needs: tissue, cell_supertype, platform, param, direction, library, term, fdr
R = pd.read_csv(TABS/"flagship_strata_ranking.csv")         # used to pick flagship strata

# --- knobs
LIBS     = ["Reactome_2022"]  # also add "GO_Biological_Process_2023" if desired
K_TOP    = 20                  # rows (terms) per heatmap
K_BARS   = 8                   # terms per subpanel in bar figure
FDR_CAP  = 6.0                 # cap for −log10(FDR) color scales
PLAT_ABBR = {"10x": "10x", "Smart-seq2": "SS2"}
COL_ORDER = [(p, prm, d) for p in ["10x","Smart-seq2"] for prm in ["a","b"] for d in ["up","down"]]

# ---------- helpers
def wrap_labels(strings, width=38):
    return [textwrap.fill(str(s), width=width, break_long_words=False, break_on_hyphens=False) for s in strings]

def _best_per_term(df):
    """Best (smallest) FDR per (term, platform, param, direction)."""
    return (df.sort_values(["term","platform","param","direction","fdr"])
              .drop_duplicates(["term","platform","param","direction"]))

# ---------- HEATMAP (per stratum)
def make_topk_heatmap(df, tissue, cell, library, k=20, vmax=6.0):
    sub = df.query("tissue==@tissue & cell_supertype==@cell & library==@library").copy()
    if sub.empty:
        print(f"[heatmap skip] No rows for {tissue} — {cell} — {library}")
        return

    sub = _best_per_term(sub)
    sub["col"] = sub["platform"].map(PLAT_ABBR) + "_" + sub["param"] + "_" + sub["direction"]
    mat = sub.pivot(index="term", columns="col", values="fdr")

    # keep terms with at least one sig hit
    sig_mask = (mat < 0.05).any(axis=1)
    mat = mat.loc[sig_mask]
    if mat.empty:
        print(f"[heatmap skip] No significant terms for {tissue} — {cell} — {library}")
        return

    X = (-np.log10(mat)).replace([np.inf, -np.inf], np.nan)
    X = X.assign(row_best=lambda d: d.max(axis=1, skipna=True)).sort_values("row_best", ascending=False).drop(columns="row_best")
    X = X.head(k)

    ordered_cols = [f"{PLAT_ABBR[p]}_{prm}_{d}" for (p,prm,d) in COL_ORDER if f"{PLAT_ABBR[p]}_{prm}_{d}" in X.columns]
    X = X[ordered_cols]
    plot_vals = X.fillna(0).clip(upper=vmax)

    n_terms = plot_vals.shape[0]
    fig_h = min(0.32*n_terms + 2.0, 14)
    fig_w = max(6, 0.9*plot_vals.shape[1] + 3)

    fig, ax = plt.subplots(figsize=(fig_w, fig_h), dpi=300)
    im = ax.imshow(plot_vals.values, aspect="auto", interpolation="nearest", cmap="viridis", vmin=0, vmax=vmax)

    ax.set_yticks(range(n_terms))
    ax.set_yticklabels(wrap_labels(plot_vals.index, width=42), fontsize=8)

    def pretty(c):
        plat, prm, direction = c.split("_")
        arrow = "↑" if direction=="up" else "↓"
        return f"{plat} {prm}{arrow}"

    ax.set_xticks(range(plot_vals.shape[1]))
    ax.set_xticklabels([pretty(c) for c in plot_vals.columns], rotation=45, ha="right", fontsize=9)

    ax.set_title(f"{library.replace('_',' ')} — {tissue} — {cell}\nTop {k} enriched terms (−log10 FDR, cap={vmax})", fontsize=12, pad=10)
    cbar = plt.colorbar(im, ax=ax, pad=0.01)
    cbar.set_label("−log10(FDR)", rotation=90)
    for s in ("top","right"): ax.spines[s].set_visible(False)

    fig.tight_layout()
    out = FIGS/f"topk_{library}_{tissue}_{cell}.png"
    fig.savefig(out, bbox_inches="tight")
    plt.close(fig)
    print("Saved:", out)

# ---------- BAR PANEL (per stratum)
def bar_panel_topk(df, tissue, cell, library, k=8, vmax=6.0):
    # build four subpanels: a↑, a↓, b↑, b↓
    combos = [("a","up"), ("a","down"), ("b","up"), ("b","down")]
    sub_all = df.query("tissue==@tissue & cell_supertype==@cell & library==@library").copy()
    if sub_all.empty:
        print(f"[bars skip] No rows for {tissue} — {cell} — {library}")
        return

    # find global x-limit across panels
    panel_tables = []
    for prm, direc in combos:
        sub = sub_all.query("param==@prm & direction==@direc")
        if sub.empty:
            panel_tables.append(None); continue
        sub = _best_per_term(sub)

        # choose top-k by best (min FDR) across platforms
        best_by_term = (sub.groupby("term")["fdr"].min().sort_values().head(k))
        keep_terms = list(best_by_term.index)

        # build 2-col table (10x, SS2) of −log10(FDR)
        small = sub[sub["term"].isin(keep_terms)].copy()
        small["plat"] = small["platform"].map(PLAT_ABBR)
        mat = small.pivot_table(index="term", columns="plat", values="fdr", aggfunc="min")
        mat = (-np.log10(mat)).replace([np.inf, -np.inf], np.nan).fillna(0).clip(upper=vmax)
        # order by max value
        mat = mat.assign(_ord=lambda d: d.max(axis=1)).sort_values("_ord", ascending=False).drop(columns="_ord")
        panel_tables.append(mat)

    # figure size scales with #rows in the largest panel
    max_rows = max((t.shape[0] for t in panel_tables if t is not None), default=0)
    if max_rows == 0:
        print(f"[bars skip] No significant terms for {tissue} — {cell} — {library}")
        return

    fig_h = 2 + 0.35*max_rows*2      # two rows of panels
    fig_w = 12
    fig, axes = plt.subplots(2, 2, figsize=(fig_w, fig_h), dpi=300)
    axes = axes.flatten()

    # consistent x-range
    global_max = max((np.nanmax(t.values) for t in panel_tables if t is not None), default=1.0)
    xlim = (0, max(vmax, global_max + 0.5))
    x_ref = -np.log10(0.05)  # ~1.301

    titles = [r"a $\uparrow$", r"a $\downarrow$", r"b $\uparrow$", r"b $\downarrow$"]

    for ax, mat, title in zip(axes, panel_tables, titles):
        ax.set_title(title, fontsize=11)
        if mat is None or mat.empty:
            ax.axis("off")
            continue

        terms = wrap_labels(mat.index, width=36)
        y = np.arange(len(terms))
        bar_h = 0.36

        # two bars per term if both columns exist
        cols = [c for c in ["10x","SS2"] if c in mat.columns]
        offset = -bar_h/2 if len(cols)==2 else 0.0

        for i, c in enumerate(cols):
            ax.barh(y + (offset + i*bar_h), mat[c].values, height=bar_h, label=c)
        ax.set_yticks(y)
        ax.set_yticklabels(terms, fontsize=8)
        ax.invert_yaxis()

        ax.set_xlim(xlim)
        ax.axvline(x_ref, color="k", ls="--", lw=0.8, alpha=0.6)  # FDR=0.05 reference
        ax.set_xlabel("−log10(FDR)")

        for s in ("top","right"): ax.spines[s].set_visible(False)
        ax.grid(axis="x", alpha=0.15)

    # shared legend
    handles, labels = axes[0].get_legend_handles_labels()
    if handles:
        fig.legend(handles, labels, loc="upper center", ncol=2, frameon=False, bbox_to_anchor=(0.5, 1.02))

    fig.suptitle(f"{library.replace('_',' ')} — {tissue} — {cell}\nTop {K_BARS} terms per subpanel", fontsize=13, y=1.04)
    fig.tight_layout()
    out = FIGS/f"topk_bars_{library}_{tissue}_{cell}.png"
    fig.savefig(out, bbox_inches="tight")
    plt.close(fig)
    print("Saved:", out)

# ---------- run for flagship strata
flagships = (R.sort_values("flagship_score", ascending=False)
               .head(6)[["tissue","cell_supertype"]]
               .drop_duplicates())

for lib in LIBS:
    for _, r in flagships.iterrows():
        make_topk_heatmap(E, r.tissue, r.cell_supertype, library=lib, k=K_TOP, vmax=FDR_CAP)
        bar_panel_topk(E, r.tissue, r.cell_supertype, library=lib, k=K_BARS, vmax=FDR_CAP)
```

```python
# -*- coding: utf-8 -*-
import pandas as pd, numpy as np, matplotlib.pyplot as plt, textwrap
from pathlib import Path
plt.rcParams.update({
    "font.size": 11,
    "axes.titlesize": 14,
    "axes.labelsize": 12,
    "xtick.labelsize": 10,
    "ytick.labelsize": 10,
    "figure.dpi": 150
})

ROOT = Path("/Users/david-johnlonge/Documents/Uni stuff/diss/8273102")
TABS = ROOT/"results/tables"
FIGS = ROOT/"results/figures"
(FIGS/"panels_topk").mkdir(parents=True, exist_ok=True)

# ---- Inputs
E = pd.read_csv(TABS/"enrichment/enrichr_all.csv")  # has: tissue, cell_supertype, platform, param, direction, library, term, fdr
R = pd.read_csv(TABS/"flagship_strata_ranking.csv") # choose which strata to visualize

FLAGSHIPS = (R.sort_values("flagship_score", ascending=False)
               .head(6)[["tissue","cell_supertype"]]
               .drop_duplicates())
LIBRARY = "Reactome_2022"   # change to "GO_Biological_Process_2023" if desired
TOPK = 8
THRESH = 0.05
NEGLOG10 = -np.log10(THRESH)  # ≈ 1.301

# ---------------- utilities ----------------
def wrap(s, width=42):
    """Wrap long y-labels so they don't collide; keep Reactome code on same line."""
    s = str(s)
    # prefer breaking before the R-HSA code if present
    if " R-HSA-" in s:
        name, code = s.split(" R-HSA-", 1)
        name = textwrap.fill(name.strip(), width=width)
        return f"{name}  R-HSA-{code}"
    return textwrap.fill(s, width=width)

def collect_top_terms(df, t, c, prm, direc, k=TOPK):
    """Union top-k terms across platforms by min FDR, keep only FDR<0.05 somewhere."""
    sub = df.query("tissue==@t & cell_supertype==@c & param==@prm & direction==@direc & library==@LIBRARY").copy()
    if sub.empty:
        return pd.DataFrame(columns=["term","platform","fdr"])
    # best FDR per (term, platform)
    sub = sub.sort_values(["term","platform","fdr"]).drop_duplicates(["term","platform"])
    # min FDR across platforms for ranking
    ranker = sub.groupby("term", as_index=False)["fdr"].min().rename(columns={"fdr":"min_fdr"})
    ranker = ranker[ranker["min_fdr"] < THRESH].sort_values("min_fdr").head(k)
    keep_terms = set(ranker["term"])
    return sub[sub["term"].isin(keep_terms)]

def to_matrix(best):
    """Wide matrix with index=wrapped term, columns=['10x','Smart-seq2'], values=−log10(FDR); NaNs allowed."""
    if best.empty:
        return pd.DataFrame(columns=["10x","Smart-seq2"])
    w = (best.assign(term=lambda d: d["term"].map(wrap))
              .pivot_table(index="term", columns="platform", values="fdr", aggfunc="min"))
    for col, alias in [("10x","10x"), ("Smart-seq2","Smart-seq2")]:
        if col not in w.columns:
            w[col] = np.nan
    w = w[["10x","Smart-seq2"]]
    return (-np.log10(w))

def dot_panel(ax, M, title):
    """Plot one subpanel. M is df with columns ['10x','Smart-seq2'] in −log10(FDR)."""
    y = np.arange(len(M))
    # light connector lines to show platform differences
    if len(M) > 0:
        ax.hlines(y, M.min(axis=1), M.max(axis=1), lw=1.2, color="0.75", zorder=1)
        ax.plot(M["10x"], y, "o", ms=6, label="10x", zorder=2)
        ax.plot(M["Smart-seq2"], y, "s", ms=6, label="SS2", zorder=3)
    ax.axvline(NEGLOG10, ls="--", lw=1.0, color="0.5")
    ax.set_yticks(y, M.index if len(M) else [])
    ax.invert_yaxis()
    ax.set_xlabel("−log10(FDR)")
    ax.set_title(title, loc="left")
    # nice x-limits
    xmax = max(NEGLOG10, float(np.nanmax(M.values)) if len(M) else NEGLOG10) + 0.3
    ax.set_xlim(0, xmax)

def make_stratum_figure(df, t, c, savepath):
    """Assemble 1–4 subpanels actually present; skip empty ones."""
    combos = [("a","up","a ↑"), ("a","down","a ↓"), ("b","up","b ↑"), ("b","down","b ↓")]
    matrices = []
    for prm, direc, label in combos:
        best = collect_top_terms(df, t, c, prm, direc, k=TOPK)
        M = to_matrix(best)
        matrices.append((label, M))
    # keep only non-empty
    matrices = [(lab, M) for lab, M in matrices if len(M) > 0]
    if len(matrices) == 0:
        print(f"[skip] {t} — {c}: no significant terms in any subpanel.")
        return

    n = len(matrices)
    ncols = 2 if n > 1 else 1
    nrows = int(np.ceil(n / ncols))

    # dynamic height: ~0.35 inch per row per subplot + margin
    rows_per_subplot = [len(M) for _, M in matrices]
    max_rows = max(rows_per_subplot)
    h_per_row = 0.35
    fig_h = nrows * (max_rows*h_per_row + 1.0)
    fig_w = 12 if ncols==2 else 8

    fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(fig_w, fig_h), squeeze=False)
    axes = axes.flatten()

    for ax in axes[n:]:
        ax.remove()  # drop unused frames

    for (label, M), ax in zip(matrices, axes[:n]):
        dot_panel(ax, M, label)

    # one shared legend
    handles, labels = axes[0].get_legend_handles_labels()
    if handles:
        fig.legend(handles, labels, loc="upper center", ncol=2, frameon=False, borderaxespad=0.5)

    fig.suptitle(f"{LIBRARY.replace('_',' ')} — {t} — {c}\nTop {TOPK} terms per subpanel", y=0.995)
    plt.tight_layout(rect=[0,0,1,0.96])
    fig.savefig(savepath, dpi=300)
    plt.close(fig)
    print("Saved:", savepath)

# ---------------- run for flagships ----------------
Esub = E[E["library"]==LIBRARY].copy()
for t, c in FLAGSHIPS.itertuples(index=False):
    safe_t = str(t).replace(" ","_")
    safe_c = str(c).replace(" ","_")
    out = FIGS/"panels_topk"/f"topk_dots_{LIBRARY}_{safe_t}_{safe_c}.png"
    make_stratum_figure(Esub, t, c, out)
```

```python
# === Reactome top-k dot panels (clean version) ===
import pandas as pd, numpy as np, matplotlib.pyplot as plt, textwrap as tw
from pathlib import Path

ROOT = Path("/Users/david-johnlonge/Documents/Uni stuff/diss/8273102")
TABS = ROOT/"results/tables"
FIGS = ROOT/"results/figures"
OUT  = FIGS/"panels_clean"
OUT.mkdir(parents=True, exist_ok=True)

E = pd.read_csv(TABS/"enrichment/enrichr_all.csv")  # your enrichment table

# ---------------- helpers ----------------
def wrap(s, width=42):
    if pd.isna(s): return ""
    # keep the R-HSA id but allow wrapping before it
    s = s.replace(" R-HSA", "\nR-HSA")
    return "\n".join(tw.wrap(s, width=width, break_long_words=False))

def sanitize(s):  # filesystem-safe
    return "".join([c if c.isalnum() or c in "-_." else "_" for c in str(s)])

def topk_df(df, K=8, fdr_col="fdr", cap=6.0):
    """Return tidy top-K rows for plotting; ranks by best (min) FDR across platforms."""
    if df.empty:
        return df
    # keep rows significant on either platform
    sig = df[df[fdr_col] < 0.05].copy()
    if sig.empty:
        return sig
    sig["best_fdr"] = sig.groupby("term")[fdr_col].transform("min")
    sig = (sig.sort_values(["best_fdr", fdr_col])
              .drop_duplicates(["term", "platform"])
              .groupby("term").head(2))  # at most two platforms per term
    # take top K terms by best_fdr (keep both platforms for each)
    keep_terms = (sig.groupby("term")["best_fdr"].min()
                    .sort_values().head(K).index)
    sig = sig[sig["term"].isin(keep_terms)].copy()
    sig["neglog10fdr"] = (-np.log10(sig[fdr_col])).clip(upper=cap)
    # friendly y labels, keep Reactome ID on second line
    sig["ylab"] = sig["term"].apply(wrap)
    return sig

def subpanel(ax, sub, title, cap=6.0):
    ax.set_xlim(0, cap)
    if sub.empty:
        ax.text(0.02, 0.5, "no significant terms", va="center", ha="left",
                transform=ax.transAxes, fontsize=9, color="0.4")
        ax.set_yticks([]); ax.set_xticks([0,1,2,3,4,5,6])
        ax.grid(True, axis="x", color="0.9", lw=0.8)
        ax.set_title(title, fontsize=11, pad=4)
        return

    # order terms by max signal across platforms
    order = (sub.groupby("ylab")["neglog10fdr"].max()
               .sort_values(ascending=True).index)
    sub = sub.set_index("ylab").loc[order].reset_index()
    # y positions per term (one row per term)
    y = np.arange(len(order))
    # x values for each platform
    tenx = sub[sub["platform"]=="10x"]
    ss2  = sub[sub["platform"]=="Smart-seq2"]
    # Build a mapping from term to values (may be missing on one platform)
    v10 = dict(zip(tenx["ylab"], tenx["neglog10fdr"]))
    vss = dict(zip(ss2["ylab"],  ss2["neglog10fdr"]))

    x10 = [v10.get(lbl, np.nan) for lbl in order]
    xss = [vss.get(lbl, np.nan) for lbl in order]

    # faint connectors (only where both exist)
    for yi, (a,b) in enumerate(zip(x10, xss)):
        if np.isfinite(a) and np.isfinite(b):
            ax.plot([a,b], [yi, yi], color="0.85", lw=1.2, zorder=1)

    # points
    ax.scatter(x10, y, s=28, label="10x",  zorder=3)         # default color
    ax.scatter(xss, y, s=28, label="SS2",  zorder=3, marker="s")  # square for SS2

    # y labels on the left
    ax.set_yticks(y)
    ax.set_yticklabels(order, fontsize=8)
    # x
    ax.set_xticks([0,1,2,3,4,5,6])
    ax.grid(True, axis="x", color="0.92", lw=0.8)
    # thresholds
    for x in (1.30103, 2.0):
        ax.axvline(x, ls="--", lw=0.9, color="0.7")
    ax.set_xlabel(r"$-\log_{10}(\mathrm{FDR})$", fontsize=10, labelpad=3)
    ax.set_title(title, fontsize=11, pad=4)

def plot_topk_dots(tissue, cell, library="Reactome_2022", K=8, cap=6.0):
    sub = E.query("library==@library & tissue==@tissue & cell_supertype==@cell").copy()
    # Build per-subpanel frames
    frames = {}
    for prm in ["a","b"]:
        for dirn in ["up","down"]:
            key = f"{prm}_{dirn}"
            frames[key] = topk_df(
                sub.query("param==@prm & direction==@dirn"),
                K=K, cap=cap
            )

    # Figure
    plt.rcParams.update({"font.size": 9})
    fig = plt.figure(figsize=(11.6, 6.4), constrained_layout=True)
    gs  = fig.add_gridspec(2, 2, width_ratios=[1,1])
    ax11 = fig.add_subplot(gs[0,0]); ax12 = fig.add_subplot(gs[0,1])
    ax21 = fig.add_subplot(gs[1,0]); ax22 = fig.add_subplot(gs[1,1])

    # titles for subpanels
    subpanel(ax11, frames["a_up"],   "a ↑", cap=cap)
    subpanel(ax12, frames["a_down"], "a ↓", cap=cap)
    subpanel(ax21, frames["b_up"],   "b ↑", cap=cap)
    subpanel(ax22, frames["b_down"], "b ↓", cap=cap)

    # unified figure title + legend
    fig.suptitle(f"Reactome 2022 — {tissue} — {cell}\nTop {K} terms per subpanel",
                 fontsize=13, y=0.98)
    # one compact legend for all axes
    handles, labels = ax12.get_legend_handles_labels()
    if not handles:  # if ax12 had no data, grab from any other
        for ax in [ax11, ax21, ax22]:
            handles, labels = ax.get_legend_handles_labels()
            if handles: break
    if handles:
        fig.legend(handles, labels, ncols=2, loc="upper center", frameon=False, fontsize=10)

    out = OUT/f"topk_dots_{sanitize(library)}_{sanitize(tissue)}_{sanitize(cell)}.png"
    fig.savefig(out, dpi=300)
    plt.close(fig)
    print("Saved:", out)

# --------- run for your main 5–6 strata ----------
flagships = [
    ("Tongue", "Keratinocyte"),
    ("Marrow", "Mononuclear phagocyte"),
    ("Bladder", "Urothelial cell"),
    ("Lung", "Smooth muscle"),
    ("Marrow", "Granulocyte"),
    ("Liver", "Hepatocyte"),
]
for t,c in flagships:
    plot_topk_dots(t, c, library="Reactome_2022", K=8, cap=6.0)
```

```python
# -------------------------------------------
# Clean top-k Reactome dot panels (2×2) per stratum
# -------------------------------------------
import pandas as pd, numpy as np, matplotlib.pyplot as plt
from pathlib import Path
import textwrap

ROOT = Path("/Users/david-johnlonge/Documents/Uni stuff/diss/8273102")
TABS = ROOT/"results/tables"
FIGS = ROOT/"results/figures/topk_dots_clean"
FIGS.mkdir(parents=True, exist_ok=True)

E = pd.read_csv(TABS/"enrichment/enrichr_all.csv")  # expects: tissue, cell_supertype, platform, param, direction, library, term, fdr
R = pd.read_csv(TABS/"flagship_strata_ranking.csv")

# choose the 5–6 'flagship' strata you’re using
flagships = (
    R.sort_values("flagship_score", ascending=False)
     .head(6)[["tissue","cell_supertype"]]
     .drop_duplicates()
)

LIB = "Reactome_2022"
PLATFORMS = ["10x", "Smart-seq2"]
PARAMS = ["a", "b"]
DIRS = ["up", "down"]
CUTS = (1.301, 2.0)   # dashed vertical lines at -log10(0.05) and -log10(0.01)

def _wrap_term(term: str, width=38) -> str:
    """Wrap long Reactome labels but preserve the trailing R-HSA id on same line."""
    # split off the final token if it is the R-HSA id
    parts = term.rsplit(" ", 1)
    if len(parts) == 2 and parts[1].startswith("R-HSA-"):
        base, rid = parts
        base_wrapped = textwrap.fill(base, width=width)
        return f"{base_wrapped} {rid}"
    return textwrap.fill(term, width=width)

def _panel_table(enrich, t, c, prm, direction, k=8):
    """Return table with terms and per-platform -log10(FDR) for a subpanel."""
    sub = (enrich.query("tissue==@t & cell_supertype==@c & library==@LIB "
                        "& param==@prm & direction==@direction")
                 .copy())
    # keep best (lowest) FDR per (term, platform)
    sub = (sub.sort_values(["term","platform","fdr"])
              .drop_duplicates(["term","platform"]))
    sub = sub[sub["fdr"].notna()]
    sub["mlog10"] = -np.log10(sub["fdr"])
    # consider only significant per platform (fdr < 0.05) for ranking
    sig = sub[sub["fdr"] < 0.05]
    if sig.empty:
        return pd.DataFrame(columns=["term"] + PLATFORMS)
    # union of terms across platforms; rank by max signal across platforms
    pivot = (sub.pivot(index="term", columns="platform", values="mlog10")
               .reindex(columns=PLATFORMS))
    # score to pick the “top k” rows (use max across platforms)
    pivot["_score"] = pivot.max(axis=1, skipna=True)
    top = (pivot.sort_values("_score", ascending=False)
                .drop(columns="_score")
                .head(k))
    top.index = [_wrap_term(t) for t in top.index]
    return top

def _draw_panel(ax, table, title, cap=6.0, show_legend=False):
    ax.set_title(title, fontsize=12, pad=6, loc="left")
    if table.empty:
        ax.text(0.5, 0.5, "no significant terms", ha="center", va="center",
                color="0.5", fontsize=11, transform=ax.transAxes)
        ax.set_xlim(0, cap); ax.set_ylim(0, 1)
        ax.set_yticks([]); ax.set_xticks(np.arange(0, cap+0.001, 1))
        ax.grid(axis="x", color="0.9", lw=0.8)
        for cut in CUTS: ax.axvline(cut, ls="--", color="0.7", lw=1)
        return

    plot = table.copy()
    # clip to cap to keep scales comparable
    plot = plot.clip(upper=cap)
    ys = np.arange(len(plot))[::-1]  # top to bottom
    ax.set_yticks(ys, plot.index, fontsize=9)
    # scatter for each platform, small horizontal line for visual anchor
    colors = {"10x":"C0", "Smart-seq2":"C1"}
    for j, p in enumerate(PLATFORMS):
        if p in plot.columns:
            ax.scatter(plot[p].values, ys + (0.12 if j==0 else -0.12),
                       s=30, color=colors[p], zorder=3, label=p if show_legend else None)
    # thin grey hairlines from 0 to max point to improve readability
    x_max = np.nanmax(plot.values)
    for y, row in zip(ys, plot.values):
        m = np.nanmax(row)
        if np.isfinite(m):
            ax.hlines(y, xmin=0, xmax=m, color="0.85", lw=0.8, zorder=1)

    ax.set_xlim(0, cap)
    ax.set_xlabel("")  # we’ll add a common xlabel
    ax.set_xticks(np.arange(0, cap+0.001, 1))
    ax.grid(axis="x", color="0.93", lw=0.8)
    for cut in CUTS: ax.axvline(cut, ls="--", color="0.7", lw=1)
    ax.invert_yaxis()
    for spine in ["top","right","left","bottom"]:
        ax.spines[spine].set_visible(False)

def make_topk_dot_panels(tissue, cell, k=8, cap=6.0):
    fig, axs = plt.subplots(2, 2, figsize=(14, 9), constrained_layout=True)
    panel_map = {("a","up"): (0,0), ("a","down"): (0,1),
                 ("b","up"): (1,0), ("b","down"): (1,1)}
    # build and draw each quadrant
    for prm in PARAMS:
        for direction in DIRS:
            r, cidx = panel_map[(prm, direction)]
            ax = axs[r, cidx]
            tab = _panel_table(E, tissue, cell, prm, direction, k=k)
            ttl = f"{prm} {'↑' if direction=='up' else '↓'}"
            show_leg = (prm, direction) == ("a","up")
            _draw_panel(ax, tab, ttl, cap=cap, show_legend=show_leg)

    # global labels and legend
    fig.suptitle(f"Reactome 2022 — {tissue} — {cell}\nTop {k} terms per subpanel",
                 fontsize=16, y=0.995)
    fig.supxlabel("−log10(FDR)   (values > cap clipped)", fontsize=12)
    # single legend above left panel (from that panel only)
    handles, labels = axs[0,0].get_legend_handles_labels()
    if handles:
        fig.legend(handles, labels, ncols=2, frameon=False,
                   loc="upper right", bbox_to_anchor=(0.98, 0.995), fontsize=11)

    out = FIGS / f"topk_dots_{LIB}_{tissue}_{cell}.png"
    fig.savefig(out, dpi=300)
    plt.close(fig)
    print("Saved:", out)

# ---------- run for your selected flagships ----------
for t, c in flagships.itertuples(index=False):
    make_topk_dot_panels(t, c, k=8, cap=6.0)
```

```python
# --- drop-in replacement: cleaner 2×2 dot-plot for one (tissue, cell_supertype) ---

import numpy as np
import matplotlib.pyplot as plt
from matplotlib import gridspec

def plot_topk_dots_clean(df_sub, title, outpath, k=8, cap=6.0,
                         platforms=("10x", "Smart-seq2"),
                         panels=(("a","up"), ("a","down"), ("b","up"), ("b","down"))):
    """
    df_sub: enrichment rows already filtered to a single (tissue, cell_supertype, library).
            Expected cols: ['platform','param','direction','term','fdr'] at minimum.
    """

    # ---------- small helpers ----------
    def _neglog10(x):
        x = np.asarray(x, dtype=float)
        with np.errstate(divide='ignore'):
            v = -np.log10(x)
        v[~np.isfinite(v)] = np.nan
        return np.clip(v, 0, cap)  # clamp for plotting

    def _top_terms_for_panel(param, direction):
        qq = df_sub[(df_sub["param"]==param) & (df_sub["direction"]==direction)].copy()
        if qq.empty:
            return []
        # best FDR per (term, platform)
        qq = (qq.sort_values(["term","platform","fdr"])
                .drop_duplicates(["term","platform"]))
        # union across platforms ranked by the *best* FDR
        best_any = (qq.groupby("term", as_index=False)["fdr"].min()
                      .sort_values("fdr")
                      .head(k)["term"].tolist())
        return best_any

    # ---------- figure skeleton (only margins/hspace changed) ----------
    fig = plt.figure(figsize=(12, 8))                # same overall look
    gs  = gridspec.GridSpec(2, 2,
                            left=0.34, right=0.98,   # more room for long labels
                            bottom=0.10, top=0.90,
                            hspace=0.45, wspace=0.35)

    ax = np.array([fig.add_subplot(gs[0,0]),
                   fig.add_subplot(gs[0,1]),
                   fig.add_subplot(gs[1,0]),
                   fig.add_subplot(gs[1,1])])

    # global title + legend placement that won’t collide with panels
    su = fig.suptitle(title, fontsize=14, y=0.94)
    # small legend across top center
    handles = [plt.Line2D([0],[0], marker='o', ls='',
                          mfc='C0', mec='C0', label=platforms[0]),
               plt.Line2D([0],[0], marker='s', ls='',
                          mfc='C1', mec='C1', label=platforms[1])]
    fig.legend(handles=handles, loc='upper center', frameon=False, ncol=2, bbox_to_anchor=(0.5, 0.905))

    fig.supxlabel(r"$-\log_{10}(\mathrm{FDR})$  (cap = %g)" % cap, fontsize=11, y=0.05)

    # dashed reference lines: FDR 0.05 and 0.01
    ref_lines = [1.3010, 2.0000]

    # ---------- draw each panel ----------
    for i, (param, direction) in enumerate(panels):
        ax_i = ax[i]
        # short, left-aligned panel label so it never collides
        label = f"{param} " + ("↑" if direction=="up" else "↓")
        ax_i.set_title(label, loc="left", fontsize=11, pad=2)

        terms = _top_terms_for_panel(param, direction)
        if not terms:  # clean empty panel
            ax_i.set_xlim(0, cap)
            ax_i.set_ylim(0, 1)
            for rl in ref_lines:
                ax_i.axvline(rl, color='0.85', ls='--', lw=1)
            ax_i.text(0.5, 0.5, "no significant terms",
                      transform=ax_i.transAxes, ha="center", va="center",
                      fontsize=10, color='0.45')
            ax_i.set_yticks([])
            ax_i.tick_params(axis='x', labelsize=10)
            for spine in ("top","right"):
                ax_i.spines[spine].set_visible(False)
            continue

        y = np.arange(len(terms))
        # tidy y tick labels (term + stable identifier if you keep it in the term)
        ax_i.set_yticks(y)
        ax_i.set_yticklabels(terms, fontsize=9)

        # light grid for readability
        ax_i.grid(axis='y', color='0.92', lw=0.8)
        for rl in ref_lines:
            ax_i.axvline(rl, color='0.85', ls='--', lw=1)

        # plot each platform’s dot; thin connector for paired terms
        for j, (plat, m, col) in enumerate(zip(platforms, ['o','s'], ['C0','C1'])):
            qq = df_sub[(df_sub["param"]==param) &
                        (df_sub["direction"]==direction) &
                        (df_sub["platform"]==plat)]
            vals = []
            for t in terms:
                fdr = np.nan
                if not qq.empty:
                    hit = qq[qq["term"]==t]
                    if not hit.empty:
                        fdr = float(hit["fdr"].min())
                vals.append(fdr)
            x = _neglog10(vals)
            ax_i.scatter(x, y, s=36, marker=m, c=col, edgecolors=col, zorder=3)

            # store for connectors (only if both platforms exist)
            if j == 0:
                x0 = x.copy()
            else:
                # line between platform points where both are finite
                mask = np.isfinite(x0) & np.isfinite(x)
                if mask.any():
                    for yy, x_a, x_b in zip(y[mask], x0[mask], x[mask]):
                        ax_i.plot([x_a, x_b], [yy, yy],
                                  color='0.80', lw=1, zorder=2)

        ax_i.set_xlim(0, cap)
        ax_i.tick_params(axis='x', labelsize=10)

        # clean spines
        for spine in ("top","right"):
            ax_i.spines[spine].set_visible(False)

    plt.savefig(outpath, dpi=300)
    plt.close()
```

```python
# --- imports ---
import pandas as pd, numpy as np, matplotlib.pyplot as plt
from textwrap import shorten, fill
from matplotlib.gridspec import GridSpec

# --- small helpers ---
def _short(term, keep_id=True, max_chars=42):
    """
    Shorten a Reactome term for plotting.
    Keeps the R-HSA id (after the last space) but wraps the title.
    """
    if not isinstance(term, str):
        return term
    # Split title and ID (assumes '... R-HSA-xxxxx' at the end)
    parts = term.rsplit(" ", 1)
    if len(parts) == 2 and parts[1].startswith("R-HSA-"):
        title, rid = parts
    else:
        title, rid = term, ""
    title = shorten(title, width=max_chars, placeholder="…")
    title = fill(title, width=28)  # wrap to 2–3 lines
    return f"{title} {rid}" if (keep_id and rid) else title

def _prep_panel(df, tissue, cell, panel_key, K=8, cap=6.0):
    """
    df must already be filtered to Reactome_2022 and contain:
      tissue, cell_supertype, platform, param, direction, term, fdr
    panel_key is a tuple like ('a','up') or ('b','down')
    """
    p, d = panel_key
    sub = (df[(df["tissue"]==tissue) &
              (df["cell_supertype"]==cell) &
              (df["param"]==p) &
              (df["direction"]==d)]
             .copy())
    if sub.empty:
        return pd.DataFrame(columns=["platform","term","neglog10FDR"])
    # best (lowest FDR) per (term, platform)
    sub["neglog10FDR"] = -np.log10(sub["fdr"].clip(lower=1e-300))
    best = (sub.sort_values(["term","platform","fdr"])
                .drop_duplicates(["term","platform"]))
    # rank terms by the best across platforms, take top K
    order = (best.groupby("term")["neglog10FDR"].max()
                 .sort_values(ascending=False)
                 .head(K)).index
    best = best[best["term"].isin(order)]
    # keep plotting order (largest at bottom looks nice)
    best["term"] = pd.Categorical(best["term"], categories=list(order)[::-1], ordered=True)
    best = best.sort_values(["term","platform"])
    # clamp huge values to cap so they don't blow out axes
    best["neglog10FDR"] = best["neglog10FDR"].clip(upper=cap)
    best["term_plot"] = best["term"].astype(str).map(_short)
    return best

def plot_reactome_topk_compact(df, tissue, cell, K=8, cap=6.0, outpath=None):
    """
    Four panels (a↑, a↓, b↑, b↓) with wrapped labels, one legend, constrained layout.
    """
    panels = {("a","up"):"a ↑", ("a","down"):"a ↓",
              ("b","up"):"b ↑", ("b","down"):"b ↓"}
    data = {k:_prep_panel(df, tissue, cell, k, K=K, cap=cap) for k in panels}

    # dynamic figure height: ~0.42 per row per term, with a floor
    max_rows = max((d["term"].nunique() if not d.empty else 0) for d in data.values())
    h = max(6.0, 0.42*max_rows*2 + 1.0)
    fig = plt.figure(constrained_layout=True, figsize=(13, h))
    fig.set_constrained_layout_pads(w_pad=0.3, h_pad=0.25)
    gs = GridSpec(2, 2, figure=fig)
    axes = {
        ("a","up"):   fig.add_subplot(gs[0,0]),
        ("a","down"): fig.add_subplot(gs[0,1]),
        ("b","up"):   fig.add_subplot(gs[1,0]),
        ("b","down"): fig.add_subplot(gs[1,1]),
    }

    # consistent markers/colors per platform
    mstyle = {"10x":{"marker":"o", "ls":"none"},
              "Smart-seq2":{"marker":"s", "ls":"none"}}

    # plot each panel
    for i,(key,label) in enumerate(panels.items()):
        ax = axes[key]
        dd = data[key]
        ax.set_title(label, fontsize=12, pad=6)
        ax.axvline(1.3, ls="--", lw=0.8, color="0.75")
        ax.axvline(1.7, ls="--", lw=0.8, color="0.75")
        ax.set_xlim(0, cap)
        ax.grid(axis="x", color="0.92", lw=0.8)
        ax.set_axisbelow(True)

        if dd.empty:
            ax.text(0.5, 0.5, "no significant terms",
                    ha="center", va="center", color="0.5", transform=ax.transAxes)
            ax.set_yticks([])
        else:
            # y positions
            y_terms = dd["term"].cat.categories if hasattr(dd["term"], "categories") else dd["term"].unique()[::-1]
            y_pos = np.arange(len(y_terms))
            # map term -> wrapped label
            label_map = dd.drop_duplicates("term")[["term","term_plot"]].set_index("term")["term_plot"].to_dict()
            # draw per platform
            for plat in ["10x","Smart-seq2"]:
                s = dd[dd["platform"]==plat]
                if s.empty: 
                    continue
                y = s["term"].astype(str).map({str(t):i for i,t in enumerate(y_terms)})
                ax.plot(s["neglog10FDR"].values, y.values,
                        mstyle[plat]["marker"], ms=6, label=plat)
            ax.set_yticks(range(len(y_terms)))
            ax.set_yticklabels([label_map[t] for t in y_terms], fontsize=9)
        # only bottom row gets x label
        if key[0]=="b":
            ax.set_xlabel(r"$-\log_{10}(\mathrm{FDR})$ (cap = %g)"%cap)
        else:
            ax.set_xlabel("")
        # tidy
        for spine in ["top","right","left"]:
            ax.spines[spine].set_visible(False)
        ax.tick_params(axis='y', length=0)

    # put a single legend in the top-left panel
    axes[("a","up")].legend(title="", loc="upper left", frameon=False)

    fig.suptitle(f"Reactome_2022 — {tissue} — {cell}\nTop {K} terms per subpanel",
                 fontsize=14, y=1.02)

    if outpath:
        fig.savefig(outpath, dpi=300)
        plt.close(fig)
    else:
        plt.show()
```

```python
import pandas as pd
from scipy.stats import spearmanr
import numpy as np

def sign_agreement(x, y):
    return np.mean(np.sign(x) == np.sign(y))

# expected columns:
# tissue_cell, gene, log2fc_10x_a, log2fc_ss2_a, log2fc_10x_b, log2fc_ss2_b
def concordance_table(df):
    rows = []
    for tc, g in df.groupby("tissue_cell"):
        a10, a20 = g["log2fc_10x_a"], g["log2fc_ss2_a"]
        b10, b20 = g["log2fc_10x_b"], g["log2fc_ss2_b"]
        rows.append({
            "Tissue / cell": tc,
            "n genes (shared)": len(g),
            "Spearman ρ (a)": spearmanr(a10, a20, nan_policy="omit").correlation,
            "Sign agreement (a)": sign_agreement(a10, a20),
            "Spearman ρ (b)": spearmanr(b10, b20, nan_policy="omit").correlation,
            "Sign agreement (b)": sign_agreement(b10, b20),
        })
    out = pd.DataFrame(rows).sort_values("Tissue / cell")
    return out

tbl = concordance_table(merged_gene_df)
tbl.to_csv("table1_dataset_concordance_by_tissue.csv", index=False)
# tbl.to_excel("table1_dataset_concordance_by_tissue.xlsx", index=False)
```

```python
import pandas as pd, numpy as np
from pathlib import Path

# ---------- paths ----------
ROOT = Path("/Users/david-johnlonge/Documents/Uni stuff/diss/8273102")
TABS = ROOT/"results"/"tables"
OUT1 = TABS/"platform_replication_summary.csv"            # detailed per stratum
OUT2 = TABS/"table1_dataset_concordance_by_tissue.csv"    # cleaner "Table 1"

# ---------- load ----------
ALL = pd.read_csv(TABS/"all_strata_kinetics.csv")
ALL = ALL[ALL["platform"].isin(["10x","Smart-seq2"])].copy()

def _spearman(x, y):
    x = pd.Series(x).astype(float)
    y = pd.Series(y).astype(float)
    m = np.isfinite(x) & np.isfinite(y)
    if m.sum() < 3:
        return np.nan
    return x[m].rank().corr(y[m].rank(), method="pearson")  # Spearman without scipy

def _sign_agree(x, y):
    x = pd.Series(x).astype(float)
    y = pd.Series(y).astype(float)
    m = np.isfinite(x) & np.isfinite(y)
    if m.sum() == 0:
        return np.nan
    return float((np.sign(x[m]) == np.sign(y[m])).mean())

rows = []
for (t, c), block in ALL.groupby(["tissue","cell_supertype"]):
    tenx = (block[block.platform=="10x"]
            .loc[:, ["gene","log2FC_a","log2FC_b"]]
            .rename(columns={"log2FC_a":"a_10x","log2FC_b":"b_10x"}))
    ss2  = (block[block.platform=="Smart-seq2"]
            .loc[:, ["gene","log2FC_a","log2FC_b"]]
            .rename(columns={"log2FC_a":"a_ss2","log2FC_b":"b_ss2"}))

    if tenx.empty or ss2.empty:
        continue

    M = tenx.merge(ss2, on="gene", how="inner")
    # masks for finite pairs
    mA = np.isfinite(M["a_10x"]) & np.isfinite(M["a_ss2"])
    mB = np.isfinite(M["b_10x"]) & np.isfinite(M["b_ss2"])

    rows.append({
        "tissue": t,
        "cell_supertype": c,
        "tissue_cell": f"{t} / {c}",
        "n_10x_fin_a": int(np.isfinite(tenx["a_10x"]).sum()),
        "n_ss2_fin_a": int(np.isfinite(ss2["a_ss2"]).sum()),
        "n_overlap_a": int(mA.sum()),
        "rho_a": _spearman(M.loc[mA,"a_10x"], M.loc[mA,"a_ss2"]),
        "sign_a": _sign_agree(M.loc[mA,"a_10x"], M.loc[mA,"a_ss2"]),
        "n_10x_fin_b": int(np.isfinite(tenx["b_10x"]).sum()),
        "n_ss2_fin_b": int(np.isfinite(ss2["b_ss2"]).sum()),
        "n_overlap_b": int(mB.sum()),
        "rho_b": _spearman(M.loc[mB,"b_10x"], M.loc[mB,"b_ss2"]),
        "sign_b": _sign_agree(M.loc[mB,"b_10x"], M.loc[mB,"b_ss2"]),
    })

summary = (pd.DataFrame(rows)
           .sort_values(["tissue","cell_supertype"])
           .reset_index(drop=True))

# save the detailed summary used by your bar figure
summary.to_csv(OUT1, index=False)
print("Saved:", OUT1)

# ------- make a clean Table 1 view -------
table1 = (summary.loc[:, ["tissue_cell",
                          "n_overlap_a","rho_a","sign_a",
                          "n_overlap_b","rho_b","sign_b"]]
          .rename(columns={
              "tissue_cell": "Tissue / cell",
              "n_overlap_a": "Overlap (a)",
              "rho_a": "Spearman ρ (a)",
              "sign_a": "Sign agree. (a)",
              "n_overlap_b": "Overlap (b)",
              "rho_b": "Spearman ρ (b)",
              "sign_b": "Sign agree. (b)",
          }))

# nice rounding
table1 = table1.copy()
for col in ["Spearman ρ (a)","Sign agree. (a)","Spearman ρ (b)","Sign agree. (b)"]:
    table1[col] = table1[col].round(3)

table1.to_csv(OUT2, index=False)
print("Saved:", OUT2)

# quick peek
display_cols = ["Tissue / cell","Overlap (a)","Spearman ρ (a)","Sign agree. (a)",
                "Overlap (b)","Spearman ρ (b)","Sign agree. (b)"]
print(table1[display_cols].to_string(index=False))
```
